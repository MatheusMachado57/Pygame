```python
import pygame, time
from pygame.locals import *
from random import randint

pygame.init()

largura = 650
altura  = 400

janela = pygame.display.set_mode((largura, altura))
pygame.display.set_caption('Mudança de Cenário')

def jogo():
    
    sair = True
    fimdejogo = False
    
    pygame.time.delay(30)     # dá um delay necessário para vermos o desenho;  

    # posição no eixo x e no eixo y;
    pos_y0 = altura/2                    
    pos_x0 = 0                                          
    
    velocidade1 = 5    # quantidade de pixels que o círculo anda a cada passo;
    
    fase = 1
    inicio = True
    my_d = "Right"
    
    fundo1 = pygame.image.load('ae - Copia.PNG')
    fundo2 = pygame.image.load('donkey-kong-country-snes4.png')
    fundo3 = pygame.image.load('ae.PNG')
    

    b1 = pygame.image.load('voando 2.PNG')
    b2 = pygame.image.load('voando 2 - Copia.PNG')
    bonec = b1   
    
    while sair:
                              
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                sair = False
            
        # movimentos e limitações;    
        comandos = pygame.key.get_pressed()
        
        if comandos[pygame.K_UP] and pos_y0 > 10:
                pos_y0 -= velocidade1
                
        if comandos[pygame.K_DOWN] and pos_y0 < altura - 100:
                pos_y0 += velocidade1        
                
        if comandos[pygame.K_LEFT] and pos_x0 > 0:
                bonec = b2
                pos_x0 -= velocidade1
                my_d = "Left"
                
        if comandos[pygame.K_RIGHT] and pos_x0 < largura - 100:
                bonec = b1
                pos_x0 += velocidade1
                my_d = "Right"
                
        # invocação das figuras:
        
        # nascimento na fase 1:
        if fase == 1 and inicio == True:
            janela.blit(fundo1, (0,0))
            inicio == False
            
        # indo:    
        # fase 1 pra fase 2:        
        if  abs(pos_x0 - largura) < 200 and abs(pos_y0 - altura/2) < 200 and fase == 1 and my_d == "Right":
            janela.blit(fundo2, (0,0))
            fase += 1
            pos_x0 = 10
            
        # fase 2 para fase 3:
        if  abs(pos_x0 - largura) < 200 and abs(pos_y0 - altura/2) < 200 and fase == 2 and my_d == "Right":
            janela.blit(fundo3, (0,0))
            fase += 1
            pos_x0 = 10    
            
        # voltando:            
        #  fase 3 para fase 2:
        if  pos_x0 < 100 and abs(pos_y0 - altura/2) < 200 and fase == 3 and my_d == "Left":
            janela.blit(fundo2, (0,0))
            fase -= 1
            pos_x0 = largura - 100    
            
        #  fase 2 para fase 1:
        if  pos_x0 < 100 and abs(pos_y0 - altura/2) < 200 and fase == 2 and my_d == "Left":
            janela.blit(fundo1, (0,0))
            fase -= 1
            pos_x0 = largura - 100          
                   
        janela.blit(bonec, (pos_x0, pos_y0))          
                
        pygame.display.update()                                    # atualiza a tela

jogo()        
pygame.quit()                              
```
