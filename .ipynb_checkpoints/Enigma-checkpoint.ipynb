```python
import pygame, random
from pygame.locals import *

SCREEN_WIDTH = 1000
SCREEN_HEIGHT = 600

GROUND_WIDTH = SCREEN_WIDTH
GROUND_HEIGHT = SCREEN_HEIGHT - 30

fundo = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

vidas_p  = 0

class Player(pygame.sprite.Sprite):

    def __init__(self):
        pygame.sprite.Sprite.__init__(self)
        self.images = [pygame.image.load('azul.png').convert_alpha()]
        self.current_image = 0
        self.image = pygame.image.load('azul.png').convert_alpha()
        self.rect = self.image.get_rect()
        self.rect[0] = 10
        self.rect[1] = GROUND_HEIGHT - 100
        self.vidas_p = 1
        
    def update(self):
        self.image = self.images[self.current_image]
            
    def go_up(self):  
        for i in range(len(box)):
            if abs(self.rect[0] - box[i].rect[0]) < 50 and abs(self.rect[1] - box[i].rect[1]) < 50:
                 self.rect[1] -= 5
                    
    def go_down(self):
        for i in range(len(box)):
            if abs(self.rect[0] - box[i].rect[0]) < 50 and abs(self.rect[1] - box[i].rect[1]) < 50:
                 self.rect[1] += 5
        
    def go_on(self): 
        for i in range(len(box)):
            if abs(self.rect[0] - box[i].rect[0]) < 50 and abs(self.rect[1] - box[i].rect[1]) < 50:
                 self.rect[0] -= 5
    
    def go_back(self): 
        for i in range(len(box)):    
            if abs(self.rect[0] - box[i].rect[0]) < 50 and abs(self.rect[1] - box[i].rect[1]) < 50:
                self.rect[0] += 5
                
# Objeto inanimado: #########################################################3               
                
w =[[300, GROUND_HEIGHT - 100],
    [350, GROUND_HEIGHT - 150],
    [350, GROUND_HEIGHT - 100],
    [650, GROUND_HEIGHT - 100],
    [650, GROUND_HEIGHT - 150],
    [700, GROUND_HEIGHT - 100]] 
box = []
                
for i in range(len(w)):
    class Box(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('marrom.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('marrom.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w[i][0]
            self.rect[1] = w[i][1]

        def update(self):
            self.image = self.images[self.current_image]
            
    box.append(Box())

# Preparo das aparências do jogo: ####################################################################   
pygame.init()
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

# Jogador
player_group = pygame.sprite.Group()
player = Player()
player_group.add(player)

# Objeto Inanimado
box_group = pygame.sprite.Group()
box_group.add(box)

clock = pygame.time.Clock()    

def jogo():     
    # tempo, moedas e corações:
    font = pygame.font.SysFont('arial black', 15)

    texto2 = font.render("Coração:", True, (0,0,0), (255,255,255))
    pos_texto2 = texto2.get_rect()
    pos_texto2.center = (40,65) 
    ##########################################
    
    # mensagem de game over !!
    def texto(msg, cor):
        texto0 = font.render(msg, True, cor)
        fundo.blit(texto0, [SCREEN_WIDTH/10, SCREEN_HEIGHT/2])
    ################################
    
    # parte original    
    fimdejogo = False
    play = True
    while play:
        
        # regulador de velocidade de jogo: 
        clock.tick(50)
        #################################
        
        # caso se feche a janela:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
        #################################
        
        # game over!!
        if player.vidas_p == 0: fimdejogo = True
        if fimdejogo == True:
                fundo.fill((255,255,255))
                texto("Fim de jogo, para continuar tecle RIGHT ou LEFT para sair", (0,0,0))
                pygame.display.update()  
            
                comandos = pygame.key.get_pressed()
                
                if comandos[pygame.K_RIGHT]:
                    player.vidas_p += 1
                    jogo()
                            
                if comandos[pygame.K_LEFT]:
                            play = False
                            fimdejogo = False
        ###############
        
        #se ainda houver vidas ou continue:
        else:
    
            # comandos do jogador:    
            comandos = pygame.key.get_pressed()
            if comandos[pygame.K_UP]: player.go_up
            if comandos[pygame.k_DOWN]: player.go_down    
            if comandos[pygame.K_LEFT]:  player.go_back()
            if comandos[pygame.K_RIGHT]: player.go_on()     
    
            # imagens e atualização
            fundo.fill((0,0,0))
  
            player_group.update()
            box_group.update()
    
            box_group.draw(fundo)
            player_group.draw(fundo)
    
            ground_group.draw(fundo)
            pygame.display.update()

jogo()
pygame.quit()
```
