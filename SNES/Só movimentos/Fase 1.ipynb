```python
import pygame, random
from pygame.locals import *

SCREEN_WIDTH = 1200
SCREEN_HEIGHT = 700

fundo = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

moedas_p = 0
vidas_p  = 0

class Player(pygame.sprite.Sprite):
    def __init__(self):
        pygame.sprite.Sprite.__init__(self)
        self.images = [pygame.image.load('Aazul.png').convert_alpha()]
        self.current_image = 0
        self.image = pygame.image.load('Aazul.png').convert_alpha()
        self.rect = self.image.get_rect()
        self.rect[0] = 550
        self.rect[1] = 100
        self.fall = False
        self.vidas_p = 1
        
    def update(self):
        self.image = self.images[self.current_image]
        
        if pygame.sprite.groupcollide(player_group, fire_group, 
                                      False, 
                                      False, 
                                      pygame.sprite.collide_mask):
            self.vidas_p -= 1

        # Update height
        if (pygame.sprite.groupcollide(player_group, ground_group, 
                                       False, 
                                       False, 
                                       pygame.sprite.collide_mask) or
            pygame.sprite.groupcollide(player_group, box_group, 
                                       False, 
                                       False, 
                                       pygame.sprite.collide_mask) or
            isjump):  
            return
        
        # gravidade ou não
        if  self.rect[1] > 1000: 
            self.rect[1] -= 5
            self.fall = False
        else: 
            self.rect[1] += 5
            self.fall = True

w0 = []
for i in range(20): w0.append([i*150, 550])

ground = []

for i in range(len(w0)):
    class Ground(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('terreno.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('terreno.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w0[i][0]
            self.rect[1] = w0[i][1]
        
        def update(self): self.image = self.images[self.current_image]
        
        def go_on(self):self.rect[0] -= 5
        def go_back(self):self.rect[0] += 5
            
    ground.append(Ground())
        
# Inimigo:
w1 = []
# w1.append([800, 505])
# w1.append([450, 505]) 

croc = []
                
for i in range(len(w1)):
    class Croc(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('Averde.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('Averde.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w1[i][0]
            self.rect[1] = w1[i][1]

        def update(self):
            self.image = self.images[self.current_image] 
            pygame.sprite.groupcollide(player_group, croc_group, 
                                       False, 
                                       True, 
                                       pygame.sprite.collide_mask)
            
    croc.append(Croc())

# Objeto inanimado:            
w2 = []         
w2.append([300, 505])
w2.append([350, 455])
w2.append([350, 505])
w2.append([400, 505])
w2.append([400, 455])
w2.append([400, 405])
w2.append([450, 505])
w2.append([450, 455])
w2.append([500, 505])
box = []
                
for i in range(len(w2)):
    class Box(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('Amarrom.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('Amarrom.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w2[i][0]
            self.rect[1] = w2[i][1]

        def update(self):
            self.image = self.images[self.current_image]
            
            # Update height
            if (pygame.sprite.groupcollide(box_group, ground_group, 
                                           False, 
                                           False, 
                                           pygame.sprite.collide_mask) or
                isjump):  
                return
        
            # gravidade ou não
            if  self.rect[1] > 1000: 
                self.rect[1] -= 5
                self.fall = False
            else: 
                self.rect[1] += 5
                self.fall = True 
        
        def go_on(self):
            
            for i in range(len(w2)):
                if box[i].rect[0] - player.rect[0] <= 50 and box[i].rect[1] - player.rect[1] <= 50:
                    return
            
            self.rect[0] -= 5
            
        def go_back(self): 
            
            for i in range(len(w2)):
                if player.rect[0] - box[i].rect[0] <= 50 and box[i].rect[1] - player.rect[1] <= 50:
                    return
            
            self.rect[0] += 5
 
    box.append(Box())
    
# Moedas:  
w3 = []     
# w3.append([100, 200])
# w3.append([120, 200])
# w3.append([140, 200])      
coin = []        
        
for i in range(len(w3)):
    class Coin(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('Aamarelo.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('Aamarelo.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w3[i][0]
            self.rect[1] = w3[i][1]

        def update(self):
            self.image = self.images[self.current_image]
            if pygame.sprite.groupcollide(player_group, coin_group, 
                                          False, True, 
                                          pygame.sprite.collide_mask):
                moedas_p += 1
    coin.append(Coin())
    
# Coração:
w4 = []
# w4.append([100, 220])
# w4.append([120, 220])
# w4.append([140, 220])

coração = []

for i in range(len(w4)):
    class Coração(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('Aroxo.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('Aroxo.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w4[i][0]
            self.rect[1] = w4[i][1]

        def update(self):
            self.image = self.images[self.current_image]
            if pygame.sprite.groupcollide(player_group, coração_group, 
                                          False, True, 
                                          pygame.sprite.collide_mask):
                vidas_p += 1
            
    coração.append(Coração())
    
# Fogo:
w5 = []
# j = 400
# for i in range(10):
#     w5.append([j, 545])
#     j += 10

fire = []

for i in range(len(w5)):
    class Fire(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('Blaranja.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('Blaranja.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w5[i][0]
            self.rect[1] = w5[i][1]

        def update(self):
            self.image = self.images[self.current_image]
            pygame.sprite.groupcollide(player_group, coração_group, 
                                       False, True, 
                                       pygame.sprite.collide_mask)
            
    fire.append(Fire())

# Preparo das aparências do jogo: 
pygame.init()
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

# Jogador
player_group = pygame.sprite.Group()
player = Player()
player_group.add(player)

# Terreno
ground_group = pygame.sprite.Group()
ground_group.add(ground)

# Inimigo:
croc_group = pygame.sprite.Group()
croc_group.add(croc)

# Objeto Inanimado
box_group = pygame.sprite.Group()
box_group.add(box)

# Moedas
coin_group = pygame.sprite.Group()
coin_group.add(coin)

# Coração
coração_group = pygame.sprite.Group()
coração_group.add(coração)

# Fire
fire_group = pygame.sprite.Group()
fire_group.add(fire)

clock = pygame.time.Clock()    
isjump = False

def jogo(): 
    
    # Marcador de tempo:
    timer = 0
    tempo_segundos = 0
    
    # tempo, moedas e corações:
    font = pygame.font.SysFont('arial black', 15)

    texto1 = font.render("Coin:", True, (0,0,0), (255,255,255))
    pos_texto1 = texto1.get_rect()
    pos_texto1.center = (50,20) 

    texto2 = font.render("Life:", True, (0,0,0), (255,255,255))
    pos_texto2 = texto2.get_rect()
    pos_texto2.center = (150,20) 

    texto3 = font.render("Time:", True, (0,0,0), (255,255,255))
    pos_texto3 = texto3.get_rect()
    pos_texto3.center = (250,20)
    
    # mensagem de game over !!
    def texto(msg, cor):
        texto0 = font.render(msg, True, cor)
        fundo.blit(texto0, [SCREEN_WIDTH/3, SCREEN_HEIGHT/2])
    
    # variáveis da desgraça do pulo:
    jumpcount = 20
    v = []    
    while jumpcount >= -20:    
        if jumpcount < 0: neg = -1
        else: neg = 1
        v.append((jumpcount**2)*0.5*neg/8)
        jumpcount -= 1

    jumpcount = 20
    i = 0
    j = 0
    s = []
    isjump = False
    
    # parte original    
    fimdejogo = False
    play = True
    while play:
        
        # regulador de velocidade de jogo: 
        clock.tick(50)
        
        # caso se feche a janela:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
        
        # game over!!
        if player.vidas_p == 0: fimdejogo = True
        if fimdejogo == True:
                fundo.fill((255,255,255))
                texto("Fim de jogo, para continuar tecle -> ou tecle <- para sair", (0,0,0))
                pygame.display.update()  
            
                comandos = pygame.key.get_pressed()
                
                if comandos[pygame.K_RIGHT]:
                    player.rect[0] = 10
                    player.rect[1] = 100
                    jogo()
                            
                if comandos[pygame.K_LEFT]:
                            play = False
                            fimdejogo = False
        
        # se ainda houver vidas ou continue:
        else:
            # comandos do jogador:    
            comandos = pygame.key.get_pressed()
            if comandos[pygame.K_LEFT]:
                for i in range(len(w0)):
                    ground[i].go_back()
                for i in range(len(w2)):
                    box[i].go_back()
            
            if comandos[pygame.K_RIGHT]:
                for i in range(len(w0)):
                    ground[i].go_on()
                for i in range(len(w2)):
                    box[i].go_on()
    
            # pulo
            if (isjump == False):
                if ((comandos[pygame.K_SPACE] and 
                    (ground[i].rect[1] > player.rect[1]  or 
                     pygame.sprite.groupcollide(player_group, box_group, 
                                                False, 
                                                False, 
                                                pygame.sprite.collide_mask))) or
                     pygame.sprite.groupcollide(player_group, croc_group, 
                                                False, 
                                                True, 
                                                pygame.sprite.collide_mask)): 
                     isjump = True
            else:
                    s.append(v)
                    if jumpcount >= -20: 
                        if jumpcount < 0: neg = -1
                        else: neg = 1    
                        player.rect[1] -= s[j][i]
                        # box
                        if pygame.sprite.groupcollide(player_group, 
                                                      box_group, 
                                                      False, 
                                                      False, 
                                                      pygame.sprite.collide_mask):
                            player.rect[1] += s[0][i]
                        # chão
                        if pygame.sprite.groupcollide(player_group, 
                                                      ground_group, 
                                                      False, 
                                                      False, 
                                                      pygame.sprite.collide_mask):
                            player.rect[1] += s[0][i]
                        # crocodilo
                        if pygame.sprite.groupcollide(player_group, 
                                                      croc_group, 
                                                      False, 
                                                      True, 
                                                      pygame.sprite.collide_mask):
                                i = 0
                                s[0] = v
                                jumpcount = 20
                        jumpcount -= 1
                        i += 1
                    else: 
                        i = 0
                        s.remove(v)
                        jumpcount = 20
                        isjump = False
            # Fim do pulo
            
            # marcador de moedas
            if moedas_p == 100:
                mordas_p = 0
                player.vidas_p += 1
            # marcador de moedas: 
            texto1 = font.render("Coin:" + str(moedas_p) + "", 
                                 True, 
                                 (0,0,0), 
                                 (255,255,255))
            # marcador de vida:
            texto2 = font.render("Life:" + str(int(player.vidas_p)) + "", 
                                 True, 
                                 (0,0,0), 
                                 (255,255,255))
            # marcador de tempo:
            if(timer < 20): 
                timer += 1
            else:
                tempo_segundos += 1
            texto3 = font.render("Time:" + str(tempo_segundos) + "", 
                                 True, 
                                 (0,0,0), 
                                 (255,255,255))
            
            # imagens e atualização
            fundo.fill((127,127,127))
            fundo.blit(texto1, pos_texto1)
            fundo.blit(texto2, pos_texto2)
            fundo.blit(texto3, pos_texto3)
  
            ground_group.update()
            player_group.update()
            box_group.update()
            coin_group.update()
            coração_group.update()
            croc_group.update()
            fire_group.update()
    
            ground_group.draw(fundo)
            fire_group.draw(fundo)
            croc_group.draw(fundo)
            coração_group.draw(fundo)
            coin_group.draw(fundo)
            box_group.draw(fundo)
            player_group.draw(fundo)
            
            pygame.display.update()

jogo()
pygame.quit()
```
