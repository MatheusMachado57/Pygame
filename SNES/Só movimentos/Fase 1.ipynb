```python
import pygame, random
from pygame.locals import *

SCREEN_WIDTH = 1200
SCREEN_HEIGHT = 700

fundo = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

moedas_p = 0
vidas_p  = 0

class Player(pygame.sprite.Sprite):
    def __init__(self):
        pygame.sprite.Sprite.__init__(self)
        self.images = [pygame.image.load('Aazul.png').convert_alpha()]
        self.current_image = 0
        self.image = pygame.image.load('Aazul.png').convert_alpha()
        self.rect = self.image.get_rect()
        self.rect[0] = 550
        self.rect[1] = 100
        self.fall = False
        self.vidas_p = 1
        
    def update(self):
        self.image = self.images[self.current_image]

        # Update height
        if (pygame.sprite.groupcollide(player_group, 
                                       ground_group, 
                                       False, 
                                       False, 
                                       pygame.sprite.collide_mask) or
            pygame.sprite.groupcollide(player_group, 
                                       box_group, 
                                       False, 
                                       False, 
                                       pygame.sprite.collide_mask) or
            isjump):  
            return
        else:
            # gravidade ou não
            if isjump == False: 
                self.rect[1] += 5

# Chão:
w0 = []
for i in range(10):w0.append([i*150, 550])
ground = []
for i in range(len(w0)):
    class Ground(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('terreno.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('terreno.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w0[i][0]
            self.rect[1] = w0[i][1]
        
        def update(self): self.image = self.images[self.current_image]
        
        def go_on(self):
            if (pygame.sprite.groupcollide(box_group, 
                                           player_group, 
                                           False, 
                                           False, 
                                           pygame.sprite.collide_mask)): 
                return
                #
            self.rect[0] -= 5
                
        def go_back(self):
            if (pygame.sprite.groupcollide(box_group, 
                                           player_group, 
                                           False, 
                                           False, 
                                           pygame.sprite.collide_mask)): 
                return
                #
            self.rect[0] += 5
            
    ground.append(Ground())
        
# Objeto inanimado:            
w1 = []         
for i in range(5): w1.append([50, 505 - i*40])
for i in range(5): w1.append([800 + i*50, 505])
for i in range(3): w1.append([850 + i*50, 465])
for i in range(1): w1.append([900 + i*50, 425])
for i in range(5): w1.append([150 + 50*i, 305 - i*40])
    
box = []
                
for i in range(len(w1)):
    class Box(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('box.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('box.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w1[i][0]
            self.rect[1] = w1[i][1]

        def update(self):
            self.image = self.images[self.current_image]
            
            # Update height
            if (pygame.sprite.groupcollide(box_group, 
                                           ground_group, 
                                           False, 
                                           False, 
                                           pygame.sprite.collide_mask) or
                isjump):  
                return
            else:
                self.rect[1] += 5
        
        def go_on(self):self.rect[0] -= 5
        def go_back(self):self.rect[0] += 5
 
    box.append(Box())        
    
# Inimigo:
w2 = []
# w2.append([800, 505])
# w2.append([450, 505]) 

croc = []
                
for i in range(len(w2)):
    class Croc(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('Averde.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('Averde.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w2[i][0]
            self.rect[1] = w2[i][1]

        def update(self):
            self.image = self.images[self.current_image] 
            pygame.sprite.groupcollide(player_group, 
                                       croc_group, 
                                       False, 
                                       True, 
                                       pygame.sprite.collide_mask)
            
    croc.append(Croc())

# Preparo das aparências do jogo: 
pygame.init()
screen = pygame.display.set_mode((SCREEN_WIDTH, 
                                  SCREEN_HEIGHT))

# Jogador:
player_group = pygame.sprite.Group()
player = Player()
player_group.add(player)

# Terreno:
ground_group = pygame.sprite.Group()
ground_group.add(ground)

# Inimigo:
croc_group = pygame.sprite.Group()
croc_group.add(croc)

# Objeto Inanimado:
box_group = pygame.sprite.Group()
box_group.add(box)

clock = pygame.time.Clock()    
isjump = False

def jogo(): 
    
    # Marcador de tempo:
    timer = 0
    tempo_segundos = 0
    
    # tempo, moedas e corações:
    font = pygame.font.SysFont('arial black', 15)

    texto1 = font.render("Coin:", True, (0,0,0), (255,255,255))
    pos_texto1 = texto1.get_rect()
    pos_texto1.center = (50,20) 

    texto2 = font.render("Life:", True, (0,0,0), (255,255,255))
    pos_texto2 = texto2.get_rect()
    pos_texto2.center = (150,20) 

    texto3 = font.render("Time:", True, (0,0,0), (255,255,255))
    pos_texto3 = texto3.get_rect()
    pos_texto3.center = (250,20)
    
    # mensagem de game over !!
    def texto(msg, cor):
        texto0 = font.render(msg, True, cor)
        fundo.blit(texto0, [SCREEN_WIDTH/3, SCREEN_HEIGHT/2])
    
    # variáveis da desgraça do pulo:
    jumpcount = 20
    v = []    
    while jumpcount >= -20:    
        if jumpcount < 0: neg = -1
        else: neg = 1
        v.append((jumpcount**2)*0.5*neg/8)
        jumpcount -= 1
        
    jumpcount = 20
    i = 0
    j = 0
    s = []
    isjump = False
    
    # parte original    
    fimdejogo = False
    play = True
    while play:
        
        # regulador de velocidade de jogo: 
        clock.tick(50)
        
        # caso se feche a janela:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
        
        # game over!!
        if player.vidas_p == 0: fimdejogo = True
        if fimdejogo == True:
                fundo.fill((255,255,255))
                texto("Fim de jogo, para continuar tecle -> ou tecle <- para sair", (0,0,0))
                pygame.display.update()  
            
                comandos = pygame.key.get_pressed()
                
                if comandos[pygame.K_RIGHT]:
                    player.rect[0] = 10
                    player.rect[1] = 100
                    jogo()
                            
                if comandos[pygame.K_LEFT]:
                            play = False
                            fimdejogo = False
        
        # se ainda houver vidas ou continue:
        else:
            # comandos do jogador:    
            comandos = pygame.key.get_pressed()
                
            if comandos[pygame.K_RIGHT]:
                    if not (pygame.sprite.groupcollide(player_group, 
                                           box_group, 
                                           False, 
                                           False, 
                                           pygame.sprite.collide_mask)):
                        for k in range(len(w0)):ground[k].go_on()
                        for k in range(len(w1)):box[k].go_on()
            
            if comandos[pygame.K_LEFT]:
                    if not (pygame.sprite.groupcollide(player_group, 
                                           box_group, 
                                           False, 
                                           False, 
                                           pygame.sprite.collide_mask)):
                        for k in range(len(w0)):ground[k].go_back()
                        for k in range(len(w1)):box[k].go_back()     
    
            # jump
            # se jump = False: cerificar se as condições se mantém
            # condições: 
            #            botão barra apertado e
            #            "altura" menor ou igual a 540 ou
            #            player colidindo com box;
            # se as condições forem satisfeitas: jump = True;
            if (isjump == False):
                if ((comandos[pygame.K_SPACE] and  
                    (pygame.sprite.groupcollide(player_group, 
                                                ground_group, 
                                                False, 
                                                False, 
                                                pygame.sprite.collide_mask) or 
                     pygame.sprite.groupcollide(player_group, 
                                                box_group, 
                                                False, 
                                                False, 
                                                pygame.sprite.collide_mask)))): 
                     isjump = True
            else:
                    # se jump = True
                    # acrescentar o vetor v ao vetor de vetores s
                    # não mexer nas variávis i e j;
                    s.append(v)
                    if jumpcount >= -20: 
                        if jumpcount < 0: neg = -1
                        else: neg = 1 
                        player.rect[1] -= s[j][i]
                        
                        # se cair em alguma caixa (box)
                        if pygame.sprite.groupcollide(player_group, 
                                                      box_group, 
                                                      False, 
                                                      False, 
                                                      pygame.sprite.collide_mask):
                            player.rect[1] += s[0][i]
                        
                        # se cair no chão (ground)
                        if pygame.sprite.groupcollide(player_group, 
                                                      ground_group, 
                                                      False, 
                                                      False, 
                                                      pygame.sprite.collide_mask):
                            player.rect[1] += s[0][i]
                        
                        jumpcount -= 1
                        i += 1
                    else: 
                        i = 0
                        s.remove(v)
                        jumpcount = 20
                        isjump = False
            # Fim do pulo
            
            # imagens e atualização
            fundo.fill((127,127,127))
  
            ground_group.update()
            player_group.update()
            box_group.update()
            croc_group.update()
    
            ground_group.draw(fundo)
            croc_group.draw(fundo)
            box_group.draw(fundo)
            player_group.draw(fundo)
            
            pygame.display.update()

jogo()
pygame.quit()
```
