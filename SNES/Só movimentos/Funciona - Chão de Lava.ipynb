```python
import pygame, random
from pygame.locals import *

SCREEN_WIDTH = 1000
SCREEN_HEIGHT = 600

GROUND_WIDTH = SCREEN_WIDTH
GROUND_HEIGHT = SCREEN_HEIGHT - 30

fundo = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

moedas_p = 0
vidas_p  = 0

class Ground(pygame.sprite.Sprite):
    def __init__(self):
        pygame.sprite.Sprite.__init__(self)

        self.image = pygame.image.load('layer-5.png').convert_alpha()
        self.image = pygame.transform.scale(self.image, (SCREEN_WIDTH, SCREEN_HEIGHT))
        self.mask = pygame.mask.from_surface(self.image)
        self.rect = self.image.get_rect()

class Player(pygame.sprite.Sprite):

    def __init__(self):
        pygame.sprite.Sprite.__init__(self)
        self.images = [pygame.image.load('Aazul.png').convert_alpha()]
        self.current_image = 0
        self.image = pygame.image.load('Aazul.png').convert_alpha()
        self.rect = self.image.get_rect()
        self.rect[0] = 10
        self.rect[1] = GROUND_HEIGHT - 100
        self.fall = False
        self.vidas_p = 1
        
    def update(self):
        self.image = self.images[self.current_image]
        
        if pygame.sprite.groupcollide(player_group, fire_group, False, False, pygame.sprite.collide_mask):
            self.vidas_p -= 1
            
        if player.rect[1] - croc[0].rect[1] < 40:
            if pygame.sprite.groupcollide(player_group, croc_group, True, False, pygame.sprite.collide_mask):
                    self.vidas_p -= 1

        # Update height
        if (pygame.sprite.groupcollide(player_group, ground_group, False, False, pygame.sprite.collide_mask) or
            pygame.sprite.groupcollide(player_group, box_group, False, False, pygame.sprite.collide_mask) or
            isjump):  
            return
        
        # gravidade ou não
        if self.rect[1] > GROUND_HEIGHT - 100: 
            self.rect[1] -= 5
            self.fall = False
        else: 
            self.rect[1] += 5
            self.fall = True
            
    def go_on(self): 
        if self.rect[0] < 900: 
            self.rect[0] += 5
        
        for i in range(len(box)):
            if abs(self.rect[0] - box[i].rect[0]) < 50 and abs(self.rect[1] - box[i].rect[1]) < 40:
                 self.rect[0] -= 5
    
    def go_back(self): 
        if self.rect[0] > 0: 
            self.rect[0] -= 5
            
        for i in range(len(box)):    
            if abs(self.rect[0] - box[i].rect[0]) < 50 and abs(self.rect[1] - box[i].rect[1]) < 40:
                self.rect[0] += 5
                
# Inimigo: ##################################################################

w =[[800, GROUND_HEIGHT - 100],
    [250, GROUND_HEIGHT - 100]] 

croc = []
                
for i in range(len(w)):
    class Croc(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('Averde.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('Averde.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w[i][0]
            self.rect[1] = w[i][1]

        def update(self):
            self.image = self.images[self.current_image] 
            pygame.sprite.groupcollide(player_group, croc_group, False, True, pygame.sprite.collide_mask)
            
            
    croc.append(Croc())

# Objeto inanimado: #########################################################3               
                
w =[[300, GROUND_HEIGHT - 100],
    [350, GROUND_HEIGHT - 150],
    [350, GROUND_HEIGHT - 100],
    [600, GROUND_HEIGHT - 100],
    [650, GROUND_HEIGHT - 100],
    [650, GROUND_HEIGHT - 150],
    [700, GROUND_HEIGHT - 100]] 
box = []
                
for i in range(len(w)):
    class Box(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('Amarrom.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('Amarrom.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w[i][0]
            self.rect[1] = w[i][1]

        def update(self):
            self.image = self.images[self.current_image]
            
    box.append(Box())
    
# Moedas: #########################################################################    
        
w = [[100, 200],
     [120, 200],
     [140, 200]]      
coin = []        
        
for i in range(len(w)):
    class Coin(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('Aamarelo.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('Aamarelo.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w[i][0]
            self.rect[1] = w[i][1]

        def update(self):
            self.image = self.images[self.current_image]
            if pygame.sprite.groupcollide(player_group, coin_group, False, True, pygame.sprite.collide_mask):
                moedas_p += 1
            
            
    coin.append(Coin())
    
# Coração: ###############################################################################
        
w = [[100, 220],
     [120, 220],
     [140, 220]]
coração = []

for i in range(len(w)):
    class Coração(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('Aroxo.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('Aroxo.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w[i][0]
            self.rect[1] = w[i][1]

        def update(self):
            self.image = self.images[self.current_image]
            if pygame.sprite.groupcollide(player_group, coração_group, False, True, pygame.sprite.collide_mask):
                vidas_p += 1
            
    coração.append(Coração())
    
# Fogo: ##############################################################################################
w = []
j = 400
for i in range(5):
    w.append([j, 510])
    j += 10

fire = []

for i in range(len(w)):
    class Fire(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('Alaranja.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('Alaranja.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w[i][0]
            self.rect[1] = w[i][1]

        def update(self):
            self.image = self.images[self.current_image]
            pygame.sprite.groupcollide(player_group, coração_group, 
                                       False, True, 
                                       pygame.sprite.collide_mask)
            
    fire.append(Fire())

# Preparo das aparências do jogo: ####################################################################   
pygame.init()
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

# Terreno
ground_group = pygame.sprite.Group()
ground = Ground()
ground_group.add(ground)

# Jogador
player_group = pygame.sprite.Group()
player = Player()
player_group.add(player)

# Inimigo:
croc_group = pygame.sprite.Group()
croc_group.add(croc)

# Objeto Inanimado
box_group = pygame.sprite.Group()
box_group.add(box)

# Moedas
coin_group = pygame.sprite.Group()
coin_group.add(coin)

# Coração
coração_group = pygame.sprite.Group()
coração_group.add(coração)

# Fire
fire_group = pygame.sprite.Group()
fire_group.add(fire)

clock = pygame.time.Clock()    
isjump = False

def jogo(): 
    
    # Marcador de tempo:
    timer = 0
    tempo_segundos = 0
    ###################
    
    # tempo, moedas e corações:
    font = pygame.font.SysFont('arial black', 15)

    texto1 = font.render("=>Time:", True, (0,0,0), (255,255,255))
    pos_texto1 = texto1.get_rect()
    pos_texto1.center = (40,40) 

    texto2 = font.render("=>Life:", True, (0,0,0), (255,255,255))
    pos_texto2 = texto2.get_rect()
    pos_texto2.center = (40,65) 

    texto3 = font.render("=>Coin:", True, (0,0,0), (255,255,255))
    pos_texto3 = texto3.get_rect()
    pos_texto3.center = (40,90)
    ##########################################
    
    # mensagem de game over !!
    def texto(msg, cor):
        texto0 = font.render(msg, True, cor)
        fundo.blit(texto0, [SCREEN_WIDTH/10, SCREEN_HEIGHT/2])
    ################################
    
    # variáveis da desgraça do pulo:
    jumpcount = 20
    v = []    
    while jumpcount >= -20:    
        if jumpcount < 0: neg = -1
        else: neg = 1
        v.append((jumpcount**2)*0.5*neg/8)
        jumpcount -= 1

    jumpcount = 20
    i = 0
    j = 0
    s = []
    isjump = False
    ###############################
    
    # parte original    
    fimdejogo = False
    play = True
    while play:
        
        # regulador de velocidade de jogo: 
        clock.tick(50)
        #################################
        
        # caso se feche a janela:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
        #################################
        
        # game over!!
        if player.vidas_p == 0: fimdejogo = True
        if fimdejogo == True:
                fundo.fill((255,255,255))
                texto("Fim de jogo, para continuar tecle RIGHT ou LEFT para sair", (0,0,0))
                pygame.display.update()  
            
                comandos = pygame.key.get_pressed()
                
                if comandos[pygame.K_RIGHT]:
                    player.vidas_p += 1
                    player.rect[0] = 10
                    player.rect[1] = GROUND_HEIGHT - 105
                    jogo()
                            
                if comandos[pygame.K_LEFT]:
                            play = False
                            fimdejogo = False
        ###############
        
        #se ainda houver vidas ou continue:
        else:
    
            # comandos do jogador:    
            comandos = pygame.key.get_pressed()
            if comandos[pygame.K_LEFT]:  player.go_back()
            if comandos[pygame.K_RIGHT]: player.go_on()     
    
            # pulo
            if (isjump == False):
                if ((comandos[pygame.K_SPACE] and 
                    (ground.rect[1] - player.rect[1] < -469 or 
                     pygame.sprite.groupcollide(player_group, box_group, 
                                                False, False, 
                                                pygame.sprite.collide_mask))) or
                     pygame.sprite.groupcollide(player_group, croc_group, 
                                                False, True, 
                                                pygame.sprite.collide_mask)): 
                     isjump = True
           
            else:
                    s.append(v)
                    if jumpcount >= -20: 
                
                        if jumpcount < 0: neg = -1
                        else: neg = 1    
                
                        player.rect[1] -= s[j][i]
                
                        if pygame.sprite.groupcollide(player_group, box_group, 
                                                      False, False, 
                                                      pygame.sprite.collide_mask):
                            player.rect[1] += s[0][i]
                
                        if pygame.sprite.groupcollide(player_group, croc_group, 
                                                      False, False, 
                                                      pygame.sprite.collide_mask):
                                i = 0
                                s[0] = v
                                jumpcount = 20
                    
                        jumpcount -= 1
                        i += 1
                    else: 
                        i = 0
                        s.remove(v)
                        jumpcount = 20
                        isjump = False
            # Fim do pulo
    
            # marcador de moedas
            if moedas_p == 100:
                mordas_p = 0
                player.vidas_p += 1
    
            # marcador de tempo:
            if(timer < 20): timer += 1
            else:
                tempo_segundos += 1
                texto1 = font.render("=>Time:" + str(tempo_segundos) + "", 
                                     True, 
                                     (0,0,0), 
                                     (225,225,255))
    
            # marcador de vida:
            texto2 = font.render("=>Life:" + str(int(player.vidas_p)) + "", 
                                 True, 
                                 (0,0,0), 
                                 (225,225,255))
    
            # marcador de moedas:
            texto3 = font.render("=>Coin:" + str(moedas_p) + "", 
                                 True, 
                                 (0,0,0), 
                                 (225,225,255))
    
            # imagens e atualização
            fundo.fill((127,127,127))
            fundo.blit(texto1, pos_texto1)
            fundo.blit(texto2, pos_texto2)
            fundo.blit(texto3, pos_texto3)
  
            player_group.update()
            box_group.update()
            coin_group.update()
            coração_group.update()
            croc_group.update()
            fire_group.update()
    
            fire_group.draw(fundo)
            croc_group.draw(fundo)
            coração_group.draw(fundo)
            coin_group.draw(fundo)
            box_group.draw(fundo)
            player_group.draw(fundo)
    
            ground_group.draw(fundo)
            pygame.display.update()

jogo()
pygame.quit()
```
