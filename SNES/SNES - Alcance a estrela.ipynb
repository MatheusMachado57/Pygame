```python
import pygame
from pygame.locals import *
from random import randint

pygame.init()

janela = pygame.display.set_mode((1300, 690))           # largura x Altura
pygame.display.set_caption('Ghost')

janela_aberta = True

pos_x0 = 300                     # posição no eixo x
pos_y0 = 400                     # posição no eixo y

pos_x1 = 950 
pos_y1 = 550
pos_x2 = 1300
pos_y2 = 450
pos_x3 = 1500
pos_y3 = 350

pos_x4 = -200
pos_y4 = 500
pos_x5 = -400
pos_y5 = 400
pos_x6 = -300
pos_y6 = 300

pos_x7 = 870
pos_y7 = -100

pos_x8 = -300
pos_y8 = 100

velocidade1 = 20    # quantidade de pixels que o círculo anda a cada passo;
velocidade2 = 10
velocidade3 = 5
velocidade4 = 40

timer = 0
tempo_segundos = 0

vidas = 3

fundo = pygame.image.load('adD.PNG')

b1 = pygame.image.load('voando.PNG')
b2 = pygame.image.load('voando - Copia.PNG')
bonec = b1

fant1 = pygame.image.load('fant1 - Copia.PNG')
fant2 = pygame.image.load('fant2 - Copia.PNG')
fant3 = pygame.image.load('fant3 - Copia.PNG')

fant4 = pygame.image.load('fant1.PNG')
fant5 = pygame.image.load('fant2.PNG')
fant6 = pygame.image.load('fant3.PNG')

F1 = pygame.image.load('FantasmaEsc.PNG')
F2 = pygame.image.load('FantasmaDesp.PNG')
Big = F1

F3 = pygame.image.load('FantasmaDesp - Copia.PNG')

font = pygame.font.SysFont('arial black', 30)
texto = font.render("  Tempo: 0", True, (0,0,0), (255,255,255))
pos_texto = texto.get_rect()
pos_texto.center = (65,50) 

vida = font.render("Vidas:", True,(0,0,0),(255,255,255))

while janela_aberta:
    
        pygame.time.delay(30)     # dá um delay necessário para vermos o desenho;
        
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                janela_aberta = False
            
        # movimentos e limitações;    
        comandos = pygame.key.get_pressed()
        
        if comandos[pygame.K_UP] and pos_y0 > 10:
                pos_y0 -= velocidade1
                
        if comandos[pygame.K_DOWN] and pos_y0 < 530:
                pos_y0 += velocidade1        
                
        if comandos[pygame.K_LEFT] and pos_x0 > 0:
                bonec = b2
                pos_x0 -= velocidade1
                
        if comandos[pygame.K_RIGHT] and pos_x0 < 1200:
                bonec = b1
                pos_x0 += velocidade1                               
              
        # reaparecimento das figuras; 
        if (pos_x1  <= -400):                   # os intervalos devem ser 
            pos_x1  = randint(1100, 1200)       # diferentes para não cercar o jogador;                                                         
        if (pos_x2 <= -400):                    # o pc escolhe aleatóriamente                      
            pos_x2 = randint(1300, 1400)        # um valor intermediário;
        if (pos_x3 <= -400):
            pos_x3 = randint(1500, 1600)
            
        if (pos_x4 >= 1300): 
            pos_x4 = -int(randint(100, 200))
        if (pos_x5 >= 1300):    
            pos_x5 = -int(randint(300, 400))
        if (pos_x6 >= 1300):
            pos_x6 = -int(randint(500, 600))  
            
        if (pos_y7 > 730):
            pos_y7 = -int(randint(100, 200))            
        if (pos_x8 >= 1300):
            pos_x8 = -int(randint(150, 300))
        
        # marcador de tempo;
        if(timer < 20):
            timer += 1
        else:
            tempo_segundos +=1
            texto = font.render("  Tempo: " + str(tempo_segundos) + " ", 
                                True, (0,0,0), (225,225,255))
            timer = 0
            
        # movimentações das figuras;    
        pos_x1 -= velocidade2
        pos_x2 -= velocidade2
        pos_x3 -= velocidade2
        
        pos_x4 += velocidade2
        pos_x5 += velocidade2
        pos_x6 += velocidade2
        
        pos_y7 += velocidade3
        
        if pos_y0 < 300:
            pos_x8 += velocidade4 
         
        if pos_y7 >= 200:
            Big = F2
        else:
            Big = F1
                
        # invocação das figuras;        
        janela.blit(fundo, (0, 0))
        janela.blit(bonec, (pos_x0, pos_y0))
        
        janela.blit(fant1, (pos_x1, pos_y1))
        janela.blit(fant2, (pos_x2, pos_y2))
        janela.blit(fant3, (pos_x3, pos_y3))
        
        janela.blit(fant4, (pos_x4, pos_y4))
        janela.blit(fant5, (pos_x5, pos_y5))
        janela.blit(fant6, (pos_x6, pos_y6))
        
        janela.blit(Big, (pos_x7, pos_y7))
        janela.blit(F3,  (pos_x8, pos_y8))
        
        janela.blit(texto,  pos_texto)
        
        # detectando colisão:       
        atcf = 60
        atct = 5
        
        if (abs(pos_x0 - pos_x1) <= atcf) and (abs(pos_y0 - pos_y1) <= atcf): 
            # janela.blit(game_over,(450, 350))
            print("Game Over !!")
            janela_aberta = False 
            
        if (abs(pos_x0 - pos_x2) <= atcf) and (abs(pos_y0 - pos_y2) <= atcf): 
            # janela.blit(game_over,(450, 350))
            print("Game Over !!")
            janela_aberta = False    
            
        if (abs(pos_x0 - pos_x3) <= atcf) and (abs(pos_y0 - pos_y3) <= atcf): 
            # janela.blit(game_over,(450, 350))
            print("Game Over !!")
            janela_aberta = False  
            
        if (abs(pos_x0 - pos_x4) <= atct) and (abs(pos_y0 - pos_y4) <= atct): 
            # janela.blit(game_over,(450, 350))
            print("Game Over !!")
            janela_aberta = False
            
        if (abs(pos_x0 - pos_x5) <= atct) and (abs(pos_y0 - pos_y5) <= atct): 
            # janela.blit(game_over,(450, 350))
            print("Game Over !!")
            janela_aberta = False    
        
        if (abs(pos_x0 - pos_x6) <= atct) and (abs(pos_y0 - pos_y6) <= atct): 
            # janela.blit(game_over,(450, 350))
            print("Game Over !!")
            janela_aberta = False
        
        if (abs(pos_x0 - pos_x7) <=70) and (abs(pos_y0 - pos_y7) <=180): 
            # janela.blit(game_over,(450, 350))
            print("Game Over !!")
            janela_aberta = False
            
        if (abs(pos_x0 - pos_x8) <=180) and (abs(pos_y0 - pos_y8) <= 80): 
            # janela.blit(game_over,(450, 350))
            print("Game Over !!")
            janela_aberta = False    
            
        if (abs(pos_x0 - 1150) <= 50) and (abs(pos_y0 - 150) <= 40):
            print("Ganhou o jogo!!")
            janela_aberta = False
                   
        pygame.display.update()                                    # atualiza a tela
                
pygame.quit() 
```
