```python
import pygame, time
from pygame.locals import *
from random import randint

pygame.init()

janela = pygame.display.set_mode((1200, 600))
pygame.display.set_caption('Ghost')

def jogo():
    
    sair = True
    fimdejogo = False
    
    pygame.time.delay(30)     # dá um delay necessário para vermos o desenho;  

    # posição no eixo x e no eixo y;
    pos_y0 = 400                     
    pos_x0 = 300                                          

    pos_x1 = 950 
    pos_y1 = 550       
    
    velocidade1 = 20    # quantidade de pixels que o círculo anda a cada passo;
    velocidade2 = 10
    
    fundo = pygame.image.load('ad.PNG')

    b1 = pygame.image.load('voando 2.PNG')
    b2 = pygame.image.load('voando 2 - Copia.PNG')
    bonec = b1

    f1 = pygame.image.load('fant1 - Copia.PNG')
    f2 = pygame.image.load('fant1.PNG')
    fant1 = f1
    
    while sair:
                              
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                sair = False
            
        # movimentos e limitações;    
        comandos = pygame.key.get_pressed()
        
        if comandos[pygame.K_UP] and pos_y0 > 10:
                pos_y0 -= velocidade1
                
        if comandos[pygame.K_DOWN] and pos_y0 < 500:
                pos_y0 += velocidade1        
                
        if comandos[pygame.K_LEFT] and pos_x0 > 0:
                bonec = b2
                pos_x0 -= velocidade1
                
        if comandos[pygame.K_RIGHT] and pos_x0 < 1200:
                bonec = b1
                pos_x0 += velocidade1   
        
        #Perseguição:
        if pos_x0 - pos_x1 < 0:
            pos_x1 -= velocidade2
            fant1 = f1
            if pos_y0 == pos_y1:
                pos_x1 += velocidade2
                
        if pos_x0 - pos_x1 > 0: 
            pos_x1 += velocidade2
            fant1 = f2
            if pos_y1 == pos_y0:
                pos_x1 -= velocidade2
        
        if pos_y0 - pos_y1 < 0: 
            pos_y1 -= velocidade2
            if pos_x1 == pos_x0:
                pos_y1 += velocidade2
        
        if pos_y0 - pos_y1 > 0: 
            pos_y1 += velocidade2
            if pos_x1 == pos_x0:
                pos_y1 -= velocidade2                
                
        # invocação das figuras;        
        janela.blit(fundo, (0, 0))
        janela.blit(bonec, (pos_x0, pos_y0))        
        janela.blit(fant1, (pos_x1, pos_y1))
                   
        pygame.display.update()                                    # atualiza a tela

jogo()        
pygame.quit()                        
```
