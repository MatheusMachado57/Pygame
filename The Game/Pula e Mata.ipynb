```python
import pygame, random
from pygame.locals import *

SCREEN_WIDTH = 1000
SCREEN_HEIGHT = 600

GAME_SPEED = 10
GRAVITY = 40
SPEED = 1

GROUND_WIDTH = SCREEN_WIDTH
GROUND_HEIGHT = SCREEN_HEIGHT - 30

End1 = SCREEN_WIDTH/2 + 100
End2 = -300

class Rato(pygame.sprite.Sprite):

    def __init__(self):
        pygame.sprite.Sprite.__init__(self)

        self.images = [pygame.image.load('rato.png').convert_alpha()]

        self.current_image = 0

        self.image = pygame.image.load('rato.png').convert_alpha()
        self.mask = pygame.mask.from_surface(self.image)

        self.rect = self.image.get_rect()
        self.rect[0] = SCREEN_WIDTH / 2 - 200
        self.rect[1] = GROUND_HEIGHT - 250

    def update(self):
        self.image = self.images[ self.current_image ]

        # Update height
        if self.rect[1] < GROUND_HEIGHT - 250: self.rect[1] += GRAVITY
        
    def jump(self): 
        if self.rect[1] >= GROUND_HEIGHT - 250: 
            self.rect[1] -= 200
            self.rect[0] += 100
            
    def go_on(self): 
        if self.rect[0] < 700: self.rect[0] += 30
    
    def go_back(self): 
        if self.rect[0] > 0: self.rect[0] -= 30
            
class Snake(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('snake.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('snake.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = SCREEN_WIDTH  - 400
            self.rect[1] = GROUND_HEIGHT - 155
        
        def update(self): 
            self.image = self.images[ self.current_image ]
        
def is_off_screen(sprite): return sprite.rect[0] < -(sprite.rect[2])

pygame.init()
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

rato_group = pygame.sprite.Group()
rato = Rato()
rato_group.add(rato)

snake_group = pygame.sprite.Group()
snake = Snake()
snake_group.add(snake)

BACKGROUND = pygame.image.load('battleback6.png')
BACKGROUND = pygame.transform.scale(BACKGROUND, (SCREEN_WIDTH, SCREEN_HEIGHT))

clock = pygame.time.Clock()    

while True:
    clock.tick(10)
    
    for event in pygame.event.get():
        if event.type == QUIT:
            pygame.quit()
            
    # movimentos do pikachu
#    comandos = pygame.key.get_pressed()
#    if comandos[pygame.K_SPACE]: rato.jump()
#    if comandos[pygame.K_RIGHT]: rato.go_on()
#    if comandos[pygame.K_LEFT]:  rato.go_back()    
    
    screen.blit(BACKGROUND, (0, 0))
  
    snake_group.update()
    snake_group.draw(screen)
    
#    rato_group.update()
#    rato_group.draw(screen)
    
    pygame.display.update()
```
