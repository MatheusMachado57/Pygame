```python
import pygame, random
from pygame.locals import *

SCREEN_WIDTH = 1000
SCREEN_HEIGHT = 600

GAME_SPEED = 10
GRAVITY = 40
SPEED = 1

GROUND_WIDTH = SCREEN_WIDTH
GROUND_HEIGHT = SCREEN_HEIGHT - 30

class Pik(pygame.sprite.Sprite):

    def __init__(self):
        pygame.sprite.Sprite.__init__(self)

        self.images = [pygame.image.load('Pikstop.png').convert_alpha(),
                       pygame.image.load('Pikstop - Copia.png').convert_alpha(),
                       pygame.image.load('Pikachu 1.png').convert_alpha(),
                       pygame.image.load('Pikachu 1 - Copia.png').convert_alpha(),
                       pygame.image.load('Pikachu 4.png').convert_alpha(),
                       pygame.image.load('Pikachu 4 - Copia.png').convert_alpha()]
        
        self.image = pygame.image.load('Pikstop.png').convert_alpha()
        self.mask = pygame.mask.from_surface(self.image)

        self.rect = self.image.get_rect()
        self.rect[0] = SCREEN_WIDTH / 2 - 200
        self.rect[1] = GROUND_HEIGHT - 200

    def update(self):
        # Update height
        if self.rect[1] < GROUND_HEIGHT - 200: 
            self.rect[1] += GRAVITY
            if my_direction == "Right":
                self.image = pygame.image.load('Pikachu 1.png')
                if pygame.sprite.groupcollide(bird_group, barril_group, False, False, pygame.sprite.collide_mask):
                    self.rect[1] -= GRAVITY
                    self.image = pygame.image.load('Pikstop.png')
            else:
                self.image = pygame.image.load('Pikachu 1 - Copia.png')
                if pygame.sprite.groupcollide(bird_group, barril_group, False, False, pygame.sprite.collide_mask):
                    self.rect[1] -= GRAVITY
                    self.image = pygame.image.load('Pikstop - Copia.png')
            
            
    def stand(self):
        if self.rect[1] >= GROUND_HEIGHT - 200:
            if my_direction == "Right":
                self.image = pygame.image.load('Pikstop.png').convert_alpha()
            else:
                self.image = pygame.image.load('Pikstop - Copia.png').convert_alpha()
        
    def jump(self): 
        if self.rect[1] >= GROUND_HEIGHT - 200: 
            if my_direction == "Right":
                self.rect[1] -= 200
                self.rect[0] += 100
                self.image = pygame.image.load('Pikachu 4.png').convert_alpha()
            else:
                self.rect[1] -= 200
                self.rect[0] -= 100
                self.image = pygame.image.load('Pikachu 4 - Copia.png').convert_alpha()
            
    def go_on(self): 
        if self.rect[0] < 700: 
            self.rect[0] += 40
            if pygame.sprite.groupcollide(bird_group, barril_group, False, False, pygame.sprite.collide_mask):
                self.rect[0] -= 40
                self.image = pygame.image.load('Pikstop.png')
    
    def go_back(self): 
        if self.rect[0] > 0: 
            self.rect[0] -= 40
            if pygame.sprite.groupcollide(bird_group, barril_group, False, False, pygame.sprite.collide_mask):
                self.rect[0] += 40
                self.image = pygame.image.load('Pikstop - Copia.png')
            
class Barril1(pygame.sprite.Sprite):
    def __init__(self):
        pygame.sprite.Sprite.__init__(self)
        self.image = pygame.image.load('Barril.png').convert_alpha()
        self.mask = pygame.mask.from_surface(self.image)
        self.rect = self.image.get_rect()
        self.rect[0] = SCREEN_WIDTH - 200
        self.rect[1] = GROUND_HEIGHT - 120
        
class Barril2(pygame.sprite.Sprite):
    def __init__(self):
        pygame.sprite.Sprite.__init__(self)
        self.image = pygame.image.load('Barril.png').convert_alpha()
        self.mask = pygame.mask.from_surface(self.image)
        self.rect = self.image.get_rect()
        self.rect[0] = 100
        self.rect[1] = GROUND_HEIGHT - 120
        
pygame.init()
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

bird_group = pygame.sprite.Group()
bird = Pik()
bird_group.add(bird)

BACKGROUND = pygame.image.load('fundo.png')
BACKGROUND = pygame.transform.scale(BACKGROUND, (SCREEN_WIDTH, SCREEN_HEIGHT))

barril_group = pygame.sprite.Group()
barril = [Barril1(),Barril2()]
barril_group.add(barril)

clock = pygame.time.Clock()    

my_direction = "Right"

while True:
    clock.tick(10)
    
    for event in pygame.event.get():
        if event.type == QUIT:
            pygame.quit()
            
    # movimentos do pikachu
    comandos = pygame.key.get_pressed()
    if comandos[pygame.K_SPACE]: 
        bird.jump()
    if comandos[pygame.K_RIGHT]: 
        bird.go_on()
        my_direction = "Right"
    if comandos[pygame.K_LEFT]: 
        bird.go_back()  
        my_direction = "Left"
    
    screen.blit(BACKGROUND, (0, 0))
  
    bird_group.update()
    
    barril_group.draw(screen)
    bird_group.draw(screen)
    
    bird.stand()
    
    pygame.display.update()
```
