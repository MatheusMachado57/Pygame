```python
import pygame
from pygame.locals import *
from random import randint

pygame.init()

largura = 1300
altura  = 700

janela = pygame.display.set_mode((largura, altura))
pygame.display.set_caption('Ghost')

branco   = (255,255,255)
preto    = (0,  0,  0  )
vermelho = (255,0,  0  )
verde    = (0,  255,0  )
azul     = (0,  0,  255)

def text(msg, cor):
    texto1 = font.render(msg, True, cor)
    fundo.blit(texto1, [largura/10, altura/2])

def jogo():
    
    sair = True
    fimdejogo = False
    
    pygame.time.delay(30)     # dá um delay necessário para vermos o desenho;  

    # posição no eixo x e no eixo y;
    pos_y0 = 400                     
    pos_x0 = 300                                          

    pos_x1 = 950 
    pos_y1 = 550

    pos_x4 = -200
    pos_y4 = 500

    pos_x7 = 870
    pos_y7 = -100

    pos_x8 = -300
    pos_y8 = 100        
    
    l1 = [pos_x1]
    l2 = [pos_y1]
    
    velocidade1 = 20    # quantidade de pixels que o círculo anda a cada passo;
    velocidade2 = 10
    velocidade3 = 5
    velocidade4 = 40

    timer = 0
    tempo_segundos = 0
    
    fundo = pygame.image.load('adD.PNG')

    b1 = pygame.image.load('voando 2.PNG')
    b2 = pygame.image.load('voando 2 - Copia.PNG')
    bonec = b1

    fant1 = pygame.image.load('fant1 - Copia.PNG')

    fant4 = pygame.image.load('fant1.PNG')

    F1 = pygame.image.load('FantasmaEsc.PNG')
    F2 = pygame.image.load('FantasmaDesp.PNG')
    Big = F1

    F3 = pygame.image.load('FantasmaDesp - Copia.PNG')

    font = pygame.font.SysFont('arial black', 30)
    texto = font.render(" Tempo: 0", True, (0,0,0), (255,255,255))
    pos_texto = texto.get_rect()
    pos_texto.center = (65,50) 
    
    my_direction_v = "Up"
    my_direction_h = "Right"
    
    while sair:
        
#        while fimdejogo:
            
#            fundo.fill(branco)
#            text("Fim de jogo, para continuar tecle C ou S para sair", preto)
#            pygame.display.update()  
            
#            for event in pygame.event.get():
#                if event.type == pygame.QUIT:
#                    sair = False
#                    fimdejogo = False
#                if event.type == pygame.KEYDOWN:
#                    if event.key == pygame.K_c:
#                        jogo()
#                    if event.key == pygame.K_s:
#                        sair = False
#                        fimdejogo = False
                              
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                sair = False
            
        # movimentos e limitações;    
        comandos = pygame.key.get_pressed()
        
        if comandos[pygame.K_UP] and pos_y0 > 10:
                my_direction_v = "Up"
                pos_y0 -= velocidade1
                
        if comandos[pygame.K_DOWN] and pos_y0 < 500:
                my_direction_v = "Down"
                pos_y0 += velocidade1        
                
        if comandos[pygame.K_LEFT] and pos_x0 > 0:
                my_direction_h = "Left"
                bonec = b2
                pos_x0 -= velocidade1
                
        if comandos[pygame.K_RIGHT] and pos_x0 < 1200:
                my_direction_h = "Right"
                bonec = b1
                pos_x0 += velocidade1   
                
        # reaparecimento das figuras; 
        if (pos_x1  <= -400):                   # os intervalos devem ser diferentes para não cercar o jogador;                                                         
            pos_x1  = randint(1100, 1200)       # o pc escolhe aleatóriamente um valor intermediário;
            
        if (pos_x4 >= 1300): 
            pos_x4 = -int(randint(100, 200))
            
        if (pos_y7 > 730):
            pos_y7 = -int(randint(100, 200))            
        if (pos_x8 >= 1300):
            pos_x8 = -int(randint(150, 300))
            
        # marcador de tempo;
        if(timer < 20):
            timer += 1
        else:
            tempo_segundos +=1
            texto = font.render("  Tempo: " + str(tempo_segundos)+ " ", True, (0,0,0), (225,225,255))
            timer = 0
            
        pos_x1 -= velocidade2
#        pos_x4 += velocidade2
#        pos_y7 += velocidade3
        
#        if pos_y0 < 300:
#            pos_x8 += velocidade4 
         
        if pos_y7 >= 200:
            Big = F2
        else:
            Big = F1
            
        # invocação das figuras;        
        janela.blit(fundo, (0, 0))
        janela.blit(bonec, (pos_x0, pos_y0))
        
        janela.blit(fant1, (pos_x1, pos_y1))
        janela.blit(fant4, (pos_x4, pos_y4))
        
        janela.blit(Big, (pos_x7, pos_y7))
        janela.blit(F3,  (pos_x8, pos_y8))
        
        janela.blit(texto,  pos_texto)
        
        # detectando colisão:       
        fdf = 100
        fdt = 40 
        atct = 0
        
        # 950 e 550
        if (my_direction_h == "Right") and (abs(pos_x0 - pos_x1) <= fdf) and (abs(pos_y0 - pos_y1) <= fdf):      
            print("Game Over !!")
            sair = False
            
        if (my_direction_h == "Left" ) and (abs(pos_x0 - pos_x1) <= fdt) and (abs(pos_y0 - pos_y1) <= fdt):
            print("Game Over !!")
            sair = False
        
        if (abs(pos_x0 - pos_x4) <= atct) and (abs(pos_y0 - pos_y4) <= atct):   # -200 e 500
            # janela.blit(game_over,(450, 350))
            print("Game Over !!")
            sair = False
            
        if (abs(pos_x0 - pos_x7) <=70) and (abs(pos_y0 - pos_y7) <=180):        # 870 e -100
            # janela.blit(game_over,(450, 350))
            print("Game Over !!")
            sair = False
            
        if (abs(pos_x0 - pos_x8) <=180) and (abs(pos_y0 - pos_y8) <= 80):       # -300 e 100
            # janela.blit(game_over,(450, 350))
            print("Game Over !!")
            sair = False    
            
        if (abs(pos_x0 - 1150) <= 50) and (abs(pos_y0 - 150) <= 40):
            print("Ganhou o jogo!!")
            sair = False
                   
        pygame.display.update()                                    # atualiza a tela

jogo()        
pygame.quit()                            
```
