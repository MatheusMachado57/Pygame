```python
import pygame, random
from pygame.locals import *

SCREEN_WIDTH = 1200
SCREEN_HEIGHT = 700

fundo = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

class Player(pygame.sprite.Sprite):

    def __init__(self):
        pygame.sprite.Sprite.__init__(self)
        self.images = [pygame.image.load('AvermelhoD.png').convert_alpha(),
                       pygame.image.load('AvermelhoE.png').convert_alpha(),
                       pygame.image.load('AvermelhoC.png').convert_alpha(),
                       pygame.image.load('AvermelhoB.png').convert_alpha()]
        
        self.current_image = 0
        self.image = pygame.image.load('AvermelhoD.png').convert_alpha()
        self.rect = self.image.get_rect()
        self.rect[0] = 100
        self.rect[1] = 200
        self.vidas_p = 1
            
    def go_up(self):
        self.image = pygame.image.load('AvermelhoC.png').convert_alpha()
        for i in range(len(box)):
            if (self.rect[1]  - box[i].rect[1] == 50 and
                self.rect[0] == box[i].rect[0]): return
            
        for i in range(len(stone)):
            if (self.rect[1]  - stone[i].rect[1] == 50 and
                self.rect[0] == stone[i].rect[0]): return
        
        for i in range(len(people)):
            if (self.rect[1]  - people[0].rect[1] == 50 and
                self.rect[0] == people[0].rect[0]): return
        
        self.rect[1] -= 50
        
    def go_down(self):
        self.image = pygame.image.load('AvermelhoB.png').convert_alpha()
        for i in range(len(box)):
            if (box[i].rect[1] - self.rect[1] == 50 and
                self.rect[0] == box[i].rect[0]): return
            
        for i in range(len(stone)):
            if (stone[i].rect[1] - self.rect[1] == 50 and
                self.rect[0] == stone[i].rect[0]): return
            
        for i in range(len(people)):
            if (people[i].rect[1] - self.rect[1] == 50 and
                self.rect[0] == people[i].rect[0]): return
            
        self.rect[1] += 50
        
    def go_on(self): 
        self.image = pygame.image.load('AvermelhoD.png').convert_alpha()
        for i in range(len(box)):
            if (box[i].rect[0] - self.rect[0] == 50 and
                self.rect[1] == box[i].rect[1]): return
            
        for i in range(len(stone)):
            if (stone[i].rect[0] - self.rect[0] == 50 and
                self.rect[1] == stone[i].rect[1]): return
            
        for i in range(len(people)):
            if (people[i].rect[0] - self.rect[0] == 50 and
                self.rect[1] == people[i].rect[1]): return
           
        self.rect[0] += 50
        
    def go_back(self): 
        self.image = pygame.image.load('AvermelhoE.png').convert_alpha()
        for i in range(len(box)):
            if (self.rect[0] - box[i].rect[0] == 50 and
                self.rect[1] == box[i].rect[1]): return
            
        for i in range(len(stone)):
            if (self.rect[0] - stone[i].rect[0] == 50 and
                self.rect[1] == stone[i].rect[1]): return
            
        for i in range(len(people)):
            if (self.rect[0] - people[i].rect[0] == 50 and
                self.rect[1] == people[i].rect[1]): return
            
        self.rect[0] -= 50                 
                
# Paredes e Limitações: ##########################################             
w = []

# Horizontal:
for i in range(24):
    w.append([i*50,0])
for i in range(24):
    w.append([i*50,650])
    
# Vertical: 
for i in range(1,13):
    w.append([0,50*i])
for i in range(1,13):
    w.append([1150,50*i])
    
# Box: ##########################################################
box = []
                
for i in range(len(w)):
    class Box(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('Apreto.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('Apreto.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w[i][0]
            self.rect[1] = w[i][1]

        def update(self):
            self.image = self.images[self.current_image]
            
    box.append(Box())

# Pessoas: #####################################################
w = []
w.append([500,500]) 
w.append([400,600])
w.append([600,400])
w.append([700,300])

people = []

for i in range(len(w)):
    class People(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('AlaranjaC.png').convert_alpha(),
                           pygame.image.load('AlaranjaD.png').convert_alpha(),
                           pygame.image.load('AlaranjaE.png').convert_alpha(),
                           pygame.image.load('AlaranjaB.png').convert_alpha()]
            
            self.current_image = 0
            self.image = pygame.image.load('AlaranjaC.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w[i][0]
            self.rect[1] = w[i][1]

        def update(self):
            self.image = self.images[self.current_image]
            
    people.append(People())
    
# Pedras: #########################################################3               
w = []
w.append([200,200])
w.append([300,300])

stone = []
                
for i in range(len(w)):
    if i == 0:
        
        class Stone(pygame.sprite.Sprite):
            def __init__(self):
                pygame.sprite.Sprite.__init__(self)
                self.images = [pygame.image.load('Aazul.png').convert_alpha()]
                self.current_image = 0
                self.image = pygame.image.load('Aazul.png').convert_alpha()
                self.mask = pygame.mask.from_surface(self.image)
                self.rect = self.image.get_rect()
                self.rect[0] = w[i][0]
                self.rect[1] = w[i][1]
            
            def update(self): self.image = self.images[self.current_image]    
    else:        
        
        class Stone(pygame.sprite.Sprite):
            def __init__(self):
                pygame.sprite.Sprite.__init__(self)
                self.images = [pygame.image.load('Acinza.png').convert_alpha()]
                self.current_image = 0
                self.image = pygame.image.load('Acinza.png').convert_alpha()
                self.mask = pygame.mask.from_surface(self.image)
                self.rect = self.image.get_rect()
                self.rect[0] = w[i][0]
                self.rect[1] = w[i][1]

            def update(self): self.image = self.images[self.current_image]
            
    stone.append(Stone())

# Posição de Chegada: ##############################################
w = []
w.append([600,200])

chegada = []
    
class Chegada(pygame.sprite.Sprite):
    def __init__(self):
        pygame.sprite.Sprite.__init__(self)
        self.images = [pygame.image.load('Aamarelo.png').convert_alpha()]
        self.current_image = 0
        self.image = pygame.image.load('Aamarelo.png').convert_alpha()
        self.mask = pygame.mask.from_surface(self.image)
        self.rect = self.image.get_rect()
        self.rect[0] = w[0][0]
        self.rect[1] = w[0][1]
        
    def update(self): self.image = self.images[self.current_image]
        
chegada.append(Chegada())
        

# Preparo das aparências do jogo: #########################################################   
pygame.init()
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

# Jogador
player_group = pygame.sprite.Group()
player = Player()
player_group.add(player)

# Limitações:
box_group = pygame.sprite.Group()
box_group.add(box)

# Pessoas:
people_group = pygame.sprite.Group()
people_group.add(people)

# Pedra:
stone_group = pygame.sprite.Group()
stone_group.add(stone)

# Chegada:
chegada_group = pygame.sprite.Group()
chegada_group.add(chegada)

clock = pygame.time.Clock()    

def jogo():     
    # tempo, moedas e corações:
    font = pygame.font.SysFont('arial black', 20)

    # informações essenciais:
    texto3 = font.render(" Número de Movimentos: ", True, (255,255,255), (0,0,0))
    pos_texto3 = texto3.get_rect()
    pos_texto3.center = (200,20) 

    texto4 = font.render(" Para chutar aperte c ", True, (255,255,255), (0,0,0))
    pos_texto4 = texto4.get_rect()
    pos_texto4.center = (500,20) 

    texto5 = font.render(" Para recomeçar o jogo aperte q: ", True, (255,255,255), (0,0,0))
    pos_texto5 = texto5.get_rect()
    pos_texto5.center = (900,20)
    
    # mensagem de game over !!
    def texto1(msg, cor):
        texto0 = font.render(msg, True, cor)
        fundo.blit(texto0, [SCREEN_WIDTH/10 + 200, SCREEN_HEIGHT/2])
    
    def texto2(msg, cor):
        texto0 = font.render(msg, True, cor)
        fundo.blit(texto0, [SCREEN_WIDTH/10 + 200, SCREEN_HEIGHT/2 + 20])
    ################################
    
    # Fase:
    Fase = 0
    
    # parte original    
    fimdejogo = False
    play = True
    my_direction = "Right"
    while play:
        
        # regulador de velocidade de jogo: 
        clock.tick(10)
        #################################
        
        # caso se feche a janela:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
        #################################
        
        # game over!!
        if (stone[0].rect[0] == chegada[0].rect[0] and
            stone[0].rect[1] == chegada[0].rect[1]):  
            fimdejogo = True
            
        if fimdejogo == True:
                fundo.fill((127,127,127))
                texto1("Congratulations !!!", (0,0,0))
                texto2("A Senha para a próxima fase é: SDF", (0,0,0))
                pygame.display.update() 
            
                comandos = pygame.key.get_pressed()
                            
                if comandos[pygame.K_LEFT]:
                            play = False
                            fimdejogo = False
        ###############
        
        # se ainda houver vidas ou continue:
        else:
    
            # comandos do jogador:    
            comandos = pygame.key.get_pressed()
            if comandos[pygame.K_UP]:    
                player.go_up()
                my_direction = "Up"
            
            if comandos[pygame.K_DOWN]:  
                player.go_down()    
                my_direction = "Down"
            
            if comandos[pygame.K_LEFT]:  
                player.go_back()
                my_direction = "Left"
            
            if comandos[pygame.K_RIGHT]: 
                player.go_on()
                my_direction = "Right"
            
            if comandos[pygame.K_q]:
                pygame.quit()
                jogo()
                
            # dando o chute    
            if comandos[pygame.K_c]:  
                
                # direção do chute
                if my_direction == "Up":   
                    # qual pedra está sendo chutada 
                    for i in range(len(stone)):
                        if (player.rect[1]  - stone[i].rect[1] == 50 and            
                            player.rect[0] == stone[i].rect[0]):                  
                            p = i                                                 
                            # parede na qual pode bater
                            t = []                         
                            for k1 in range(len(box)):
                                if (box[k1].rect[0] == stone[p].rect[0] and        
                                    box[k1].rect[1]  < stone[p].rect[1]): 
                                    t.append(k1)
                            k2 = max(t)                                           
                            # a posição X de todas as pedras exceto
                            # a que está sendo chutada
                            d1 = []
                            for i in range(len(stone)):
                                if i != p : d1.append(stone[i])
                            # d1: lista de stone além de stone chutada 
                            # a posição X de todas as pedras que estão 
                            # na msm altura da pedra chutada
                            # além de terem uma posição X maior
                            d2 = []
                            if len(d1) >= 1:
                                for i in range(len(d1)):
                                    if (stone[p].rect[1]  > d1[i].rect[1] and 
                                        stone[p].rect[0] == d1[i].rect[0]): 
                                        d2.append(d1[i])
                            # pedra (se existir) ou parede?
                            if len(d2) >= 1:
                                for i in range(len(d2)):
                                    d2[i] = d2[i].rect[1]
                                d = max(d2)
                                if d > box[k2].rect[1]: k = d
                                else:                   k = box[k2].rect[1]    
                            else: k = box[k2].rect[1]
                            # distância máxima percorrida
                            for j in range(20):                                   
                                if stone[p].rect[1] - k != 50: stone[p].rect[1] -= 50
            
                # direção do chute
                if my_direction == "Down":   
                    # qual pedra está sendo chutada 
                    for i in range(len(stone)):
                        if (stone[i].rect[1] - player.rect[1] == 50 and            
                            player.rect[0] == stone[i].rect[0]):                  
                            p = i                                                 
                            # parede na qual pode bater
                            t = []                         
                            for k1 in range(len(box)):
                                if (box[k1].rect[0] == stone[p].rect[0] and        
                                    box[k1].rect[1]  > stone[p].rect[1]): t.append(k1)
                            k2 = min(t)                                           
                            # a posição X de todas as pedras exceto 
                            # a que está sendo chutada
                            d1 = []
                            for i in range(len(stone)):
                                if i != p : d1.append(stone[i])
                            # d1: lista de stone além de stone chutada 
                            # a posição X de todas as pedras que estão 
                            # na msm altura da pedra chutada
                            # além de terem uma posição X maior
                            d2 = []
                            if len(d1) >= 1:
                                for i in range(len(d1)):
                                    if (stone[p].rect[1]  < d1[i].rect[1] and 
                                        stone[p].rect[0] == d1[i].rect[0]): d2.append(d1[i])
                            # pedra (se existir) ou parede?
                            if len(d2) >= 1:
                                for i in range(len(d2)):
                                    d2[i] = d2[i].rect[1]
                                d = min(d2)
                                if d < box[k2].rect[1]: k = d
                                else:                   k = box[k2].rect[1]    
                            else: k = box[k2].rect[1]
                            # distância máxima percorrida
                            for j in range(20):                                   
                                if k - stone[p].rect[1] != 50: stone[p].rect[1] += 50
        
                # direção do chute
                if my_direction == "Right":   
                    # qual pedra está sendo chutada 
                    for i in range(len(stone)):
                        if (stone[i].rect[0] - player.rect[0] == 50 and            
                            player.rect[1] == stone[i].rect[1]):                  
                            p = i                                                 
                            # parede na qual pode bater
                            t = []                         
                            for k1 in range(len(box)):
                                if (box[k1].rect[1] == stone[p].rect[1] and        
                                    box[k1].rect[0]  > stone[p].rect[0]): t.append(k1)
                            k2 = min(t)                                           
                            # a posição X de todas as pedras 
                            # exceto a que está sendo chutada
                            d1 = []
                            for i in range(len(stone)):
                                if i != p : d1.append(stone[i])
                            # d1: lista de stone além de stone chutada 
                            # a posição X de todas as pedras que estão 
                            # na msm altura da pedra chutada
                            # além de terem uma posição X maior
                            d2 = []
                            if len(d1) >=1:
                                for i in range(len(d1)):
                                    if (stone[p].rect[0]  < d1[i].rect[0] and 
                                        stone[p].rect[1] == d1[i].rect[1]): d2.append(d1[i])
                            # pedra (se existir) ou parede?
                            if len(d2) >= 1:
                                for i in range(len(d2)):
                                    d2[i] = d2[i].rect[0]
                                d = min(d2)
                                if d < box[k2].rect[0]: k = d
                                else:                   k = box[k2].rect[0]    
                            else: k = box[k2].rect[0]
                            # distância máxima percorrida
                            for j in range(20):                                   
                                if k - stone[p].rect[0] != 50: stone[p].rect[0] += 50                        
                                    
                # direção do chute
                if my_direction == "Left":   
                    # qual pedra está sendo chutada 
                    for i in range(len(stone)):
                        if (player.rect[0]  - stone[i].rect[0] == 50 and            
                            player.rect[1] == stone[i].rect[1]):                  
                            p = i                                                 
                            # parede na qual pode bater
                            t = []                         
                            for k1 in range(len(box)):
                                if (box[k1].rect[1] == stone[p].rect[1] and        
                                    box[k1].rect[0]  < stone[p].rect[0]): t.append(k1)
                            k2 = max(t)                                           
                            # a posição X de todas as pedras exceto a que está sendo chutada
                            d1 = []
                            for i in range(len(stone)):
                                if i != p : d1.append(stone[i])
                            # d1: lista de stone além de stone chutada 
                            # a posição X de todas as pedras que estão 
                            # na msm altura da pedra chutada
                            # além de terem uma posição X maior
                            d2 = []
                            if len(d1) >=1:
                                for i in range(len(d1)):
                                    if (stone[p].rect[0]  > d1[i].rect[0] and 
                                        stone[p].rect[1] == d1[i].rect[1]): d2.append(d1[i])
                            # pedra (se existir) ou parede?
                            if len(d2) >= 1:
                                for i in range(len(d2)):
                                    d2[i] = d2[i].rect[0]
                                d = max(d2)
                                if d > box[k2].rect[0]: k = d
                                else:                   k = box[k2].rect[0]    
                            else: k = box[k2].rect[0]
                                
                            # percorrendo a distância
                            for j in range(20):                                   
                                if stone[p].rect[0] - k != 50: stone[p].rect[0] -= 50   
            
            # Pessoa vê o Player e corre em sua direção
            for i in range(len(people)):
                
                # pra cima
                if(people[i].rect[0] == player.rect[0] and 
                   people[i].rect[1] >  player.rect[1] + 50): 
                    people[i].rect[1] -= 50
                    # people[i].image = pygame.image.load('AlaranjaC.png').convert_alpha()
                
                # pra baixo
                if(people[i].rect[0] == player.rect[0] and 
                   people[i].rect[1] <  player.rect[1] - 50): 
                    people[i].rect[1] += 50
                    # people[i].image = pygame.image.load('AlaranjaB.png').convert_alpha()
            
                # left
                if(people[i].rect[1] == player.rect[1] and 
                   people[i].rect[0] >  player.rect[0] + 50): 
                    people[i].rect[0] -= 50
                    # people[i].image = pygame.image.load('AlaranjaE.png').convert_alpha()
            
                # right
                if(people[i].rect[1] == player.rect[1] and 
                   people[i].rect[0] <  player.rect[0] - 50): 
                    people[i].rect[0] += 50
                    # people[i].image = pygame.image.load('AlaranjaD.png').convert_alpha()
            
            # imagens e atualização
            fundo.fill((255,255,255))
            
            player_group.update()
            people_group.update()
            box_group.update()
            stone_group.update()
            chegada_group.update()
            
            chegada_group.draw(fundo)
            stone_group.draw(fundo)
            box_group.draw(fundo)
            people_group.draw(fundo)
            player_group.draw(fundo)
            
            fundo.blit(texto3, pos_texto3)
            fundo.blit(texto4, pos_texto4)
            fundo.blit(texto5, pos_texto5)
    
            pygame.display.update()

jogo()
pygame.quit()
```
