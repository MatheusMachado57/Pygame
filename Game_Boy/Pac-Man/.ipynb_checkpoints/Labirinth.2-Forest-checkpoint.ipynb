```python
import pygame, random
from pygame.locals import *
from random import randint

SCREEN_WIDTH = 1200
SCREEN_HEIGHT = 700

fundo = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

class Player(pygame.sprite.Sprite):

    def __init__(self):
        pygame.sprite.Sprite.__init__(self)
        self.images = [pygame.image.load('vermelhoD.png').convert_alpha(),
                       pygame.image.load('vermelhoE.png').convert_alpha(),
                       pygame.image.load('vermelhoC.png').convert_alpha(),
                       pygame.image.load('vermelhoB.png').convert_alpha()]
        
        self.current_image = 0
        self.image = pygame.image.load('vermelhoD.png').convert_alpha()
        self.rect = self.image.get_rect()
        self.rect[0] = 100
        self.rect[1] = 100
        self.vidas_p = 1
            
    def go_up(self):
        self.image = pygame.image.load('vermelhoC.png').convert_alpha()
        for i in range(len(box)):
            if (self.rect[1]  - box[i].rect[1] == 50 and
                self.rect[0] == box[i].rect[0]): return
        
        self.rect[1] -= 50
        
    def go_down(self):
        self.image = pygame.image.load('vermelhoB.png').convert_alpha()
        for i in range(len(box)):
            if (box[i].rect[1] - self.rect[1] == 50 and
                self.rect[0] == box[i].rect[0]): return
            
        self.rect[1] += 50
        
    def go_on(self): 
        self.image = pygame.image.load('vermelhoD.png').convert_alpha()
        for i in range(len(box)):
            if (box[i].rect[0] - self.rect[0] == 50 and
                self.rect[1] == box[i].rect[1]): return
           
        self.rect[0] += 50
        
    def go_back(self): 
        self.image = pygame.image.load('vermelhoE.png').convert_alpha()
        for i in range(len(box)):
            if (self.rect[0] -  box[i].rect[0] == 50 and
                self.rect[1] == box[i].rect[1]): return
            
        self.rect[0] -= 50                 
                
# Paredes e Limitações: ##########################################             
w1 = []

for i in range(13):   w1.append([i*50,0])
for i in range(14,22):w1.append([i*50,0])
for i in range(24):   w1.append([i*50,650])
for i in range(1,9): w1.append([0,50*i])
for i in range(1,9): w1.append([1150,50*i])
for i in range(1,16): w1.append([50*i,200])
for i in range(1,8):  w1.append([950, i*50])
for i in range(4,20): w1.append([50*i,400])    
for i in range(20,23): w1.append([50*i,200])
w1.append([23*50,0])

w1.append([12*50,-50])
w1.append([12*50,-100])
w1.append([13*50,-100])
w1.append([14*50,-100])
w1.append([15*50,-100])
w1.append([16*50,-100])
w1.append([17*50,-100])
w1.append([18*50,-100])
w1.append([19*50,-100])
w1.append([20*50,-100])
w1.append([21*50,-100])
w1.append([22*50,-100])
w1.append([23*50,-100])
w1.append([23*50,-50])

# Box: ##########################################################
box = []
                
for i in range(len(w1)):
    class Box(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('preto.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('preto.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w1[i][0]
            self.rect[1] = w1[i][1]

        def update(self):
            self.image = self.images[self.current_image]
            
    box.append(Box())

# Chegada: ###########################

w2 = []
w2.append([1050,300])

gold = []
    
class Gold(pygame.sprite.Sprite):
    def __init__(self):
        pygame.sprite.Sprite.__init__(self)
        self.images = [pygame.image.load('amarelo.png').convert_alpha()]
        self.current_image = 0
        self.image = pygame.image.load('amarelo.png').convert_alpha()
        self.mask = pygame.mask.from_surface(self.image)
        self.rect = self.image.get_rect()
        self.rect[0] = w2[0][0]
        self.rect[1] = w2[0][1]
        
    def update(self): self.image = self.images[self.current_image]
        
gold.append(Gold())

# Monstros: ####################

velocidade = 25

w2 = []
w2.append([1200, 450]) 
w2.append([1250, 500])
w2.append([1200, 550])
w2.append([1250, 600])
monster = []

for i in range(len(w2)):
    class Monster(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('AlaranjaE.png').convert_alpha(),
                           pygame.image.load('AlaranjaC.png').convert_alpha(),
                           pygame.image.load('AlaranjaD.png').convert_alpha(),
                           pygame.image.load('AlaranjaB.png').convert_alpha()]
            
            self.current_image = 0
            self.image = pygame.image.load('AlaranjaE.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w2[i][0]
            self.rect[1] = w2[i][1]

        def update(self):
            self.image = self.images[self.current_image]
            
    monster.append(Monster())

# Preparo das aparências do jogo: #########################################################   
pygame.init()
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

# Jogador
player_group = pygame.sprite.Group()
player = Player()
player_group.add(player)

# Limitações:
box_group = pygame.sprite.Group()
box_group.add(box)

# Chegada:
gold_group = pygame.sprite.Group()
gold_group.add(gold)

# Monstros:
monster_group = pygame.sprite.Group()
monster_group.add(monster)

clock = pygame.time.Clock()    

def jogo():     
    # tempo, moedas e corações:
    font = pygame.font.SysFont('arial black', 20)

    # informações essenciais:
    texto5 = font.render(" Para recomeçar o jogo aperte q: ", True, (255,255,255), (0,0,0))
    pos_texto5 = texto5.get_rect()
    pos_texto5.center = (900,20)
    
    # parte original    
    play = True
    while play:
        
        # regulador de velocidade de jogo: 
        clock.tick(10)

        # caso se feche a janela:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                            
        # se ainda houver vidas ou continue:
        else:
            # comandos do jogador:    
            comandos = pygame.key.get_pressed()
            
            if comandos[pygame.K_UP]:    
                player.go_up()
                my_direction = "Up"
            
            if comandos[pygame.K_DOWN]:  
                player.go_down()    
                my_direction = "Down"
            
            if comandos[pygame.K_LEFT]:  
                player.go_back()
                my_direction = "Left"
            
            if comandos[pygame.K_RIGHT]: 
                player.go_on()
                my_direction = "Right"
            
            if comandos[pygame.K_q]:
                print("U Quit!!!")
                pygame.quit()
            
            # Movimentos do Monstro
            dire = [0,0,0,0]
            for i in range(len(monster)):
                    dire[i] = randint(0,10)
                    if(dire[i] < 6): monster[i].rect[0] -= 50 
                    if monster[i].rect[0] < 0: monster[i].rect[0] = 1200
            
            # imagens e atualização
            fundo.fill((50,200,50))
            
            player_group.update()
            monster_group.update()
            box_group.update()
            gold_group.update()

            gold_group.draw(fundo)
            box_group.draw(fundo)
            monster_group.draw(fundo)
            player_group.draw(fundo)

            fundo.blit(texto5, pos_texto5)
    
            pygame.display.update()

jogo()
pygame.quit()
```
