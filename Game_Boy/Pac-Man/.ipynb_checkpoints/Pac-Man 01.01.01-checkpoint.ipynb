```python
import pygame, random
from pygame.locals import *
from random import randint

SCREEN_WIDTH = 1200
SCREEN_HEIGHT = 700

fundo = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

class Player(pygame.sprite.Sprite):

    def __init__(self):
        pygame.sprite.Sprite.__init__(self)
        self.images = [pygame.image.load('AvermelhoD.png').convert_alpha(),
                       pygame.image.load('AvermelhoE.png').convert_alpha(),
                       pygame.image.load('AvermelhoC.png').convert_alpha(),
                       pygame.image.load('AvermelhoB.png').convert_alpha()]
        
        self.current_image = 0
        self.image = pygame.image.load('AvermelhoD.png').convert_alpha()
        self.rect = self.image.get_rect()
        self.rect[0] = 100
        self.rect[1] = 200
        self.vidas_p = 1
            
    def go_up(self):
        self.image = pygame.image.load('AvermelhoC.png').convert_alpha()
        for i in range(len(box)):
            if (self.rect[1]  - box[i].rect[1] == 50 and
                self.rect[0] == box[i].rect[0]): return
        
        for i in range(len(monster)):
            if (self.rect[1]  - monster[0].rect[1] == 50 and
                self.rect[0] == monster[0].rect[0]): return
        
        self.rect[1] -= 50
        
    def go_down(self):
        self.image = pygame.image.load('AvermelhoB.png').convert_alpha()
        for i in range(len(box)):
            if (box[i].rect[1] - self.rect[1] == 50 and
                self.rect[0] == box[i].rect[0]): return
            
        for i in range(len(monster)):
            if (monster[i].rect[1] - self.rect[1] == 50 and
                self.rect[0] == monster[i].rect[0]): return
            
        self.rect[1] += 50
        
    def go_on(self): 
        self.image = pygame.image.load('AvermelhoD.png').convert_alpha()
        for i in range(len(box)):
            if (box[i].rect[0] - self.rect[0] == 50 and
                self.rect[1] == box[i].rect[1]): return
            
        for i in range(len(monster)):
            if (monster[i].rect[0] - self.rect[0] == 50 and
                self.rect[1] == monster[i].rect[1]): return
           
        self.rect[0] += 50
        
    def go_back(self): 
        self.image = pygame.image.load('AvermelhoE.png').convert_alpha()
        for i in range(len(box)):
            if (self.rect[0] -  box[i].rect[0] == 50 and
                self.rect[1] == box[i].rect[1]): return
            
        for i in range(len(monster)):
            if (self.rect[0] -  monster[i].rect[0] == 50 and
                self.rect[1] == monster[i].rect[1]): return
            
        self.rect[0] -= 50                 
                
# Paredes e Limitações: ##########################################             
w1 = []

# Horizontal:
for i in range(24):
    w1.append([i*50,0])
for i in range(24):
    w1.append([i*50,650])
    
# Vertical: 
for i in range(1,13):
    w1.append([0,50*i])
for i in range(1,13):
    w1.append([1150,50*i])

for i in range(1,5):
    w1.append([500,50*i])
for i in range(9,13):
    w1.append([500,50*i])

# Box: ##########################################################
box = []
                
for i in range(len(w1)):
    class Box(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('Apreto.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('Apreto.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w1[i][0]
            self.rect[1] = w1[i][1]

        def update(self):
            self.image = self.images[self.current_image]
            
    box.append(Box())

# Monstros: #####################################################
w2 = []
w2.append([700,100]) 
w2.append([700,200])
w2.append([700,500])
w2.append([700,600])
monster = []

for i in range(len(w2)):
    class Monster(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('AlaranjaC.png').convert_alpha(),
                           pygame.image.load('AlaranjaD.png').convert_alpha(),
                           pygame.image.load('AlaranjaE.png').convert_alpha(),
                           pygame.image.load('AlaranjaB.png').convert_alpha()]
            
            self.current_image = 0
            self.image = pygame.image.load('AlaranjaC.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w2[i][0]
            self.rect[1] = w2[i][1]

        def update(self):
            self.image = self.images[self.current_image]
            
    monster.append(Monster())

# Chegada: #########################################################3               
w3 = []
w3.append([1100,300])
star = []
                
for i in range(len(w3)):
        class Star(pygame.sprite.Sprite):
            def __init__(self):
                pygame.sprite.Sprite.__init__(self)
                self.images = [pygame.image.load('Aamarelo.png').convert_alpha()]
                self.current_image = 0
                self.image = pygame.image.load('Aamarelo.png').convert_alpha()
                self.mask = pygame.mask.from_surface(self.image)
                self.rect = self.image.get_rect()
                self.rect[0] = w3[i][0]
                self.rect[1] = w3[i][1]
            
            def update(self): self.image = self.images[self.current_image] 
            
        star.append(Star())
    
# Preparo das aparências do jogo: #########################################################   
pygame.init()
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

# Jogador
player_group = pygame.sprite.Group()
player = Player()
player_group.add(player)

# Limitações:
box_group = pygame.sprite.Group()
box_group.add(box)

# Monstros:
monster_group = pygame.sprite.Group()
monster_group.add(monster)

# Pedra:
star_group = pygame.sprite.Group()
star_group.add(star)

clock = pygame.time.Clock()    

def jogo():     
    # tempo, moedas e corações:
    font = pygame.font.SysFont('arial black', 20)

    # informações essenciais:
    texto5 = font.render(" Para recomeçar o jogo aperte q: ", True, (255,255,255), (0,0,0))
    pos_texto5 = texto5.get_rect()
    pos_texto5.center = (900,20)
    
    # parte original    
    play = True
    while play:
        
        # regulador de velocidade de jogo: 
        clock.tick(10)

        # caso se feche a janela:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
                            
        # se ainda houver vidas ou continue:
        else:
            # comandos do jogador:    
            comandos = pygame.key.get_pressed()
            
            if comandos[pygame.K_UP]:    
                player.go_up()
                my_direction = "Up"
            
            if comandos[pygame.K_DOWN]:  
                player.go_down()    
                my_direction = "Down"
            
            if comandos[pygame.K_LEFT]:  
                player.go_back()
                my_direction = "Left"
            
            if comandos[pygame.K_RIGHT]: 
                player.go_on()
                my_direction = "Right"
            
            if comandos[pygame.K_q]:
                print("U Quit!!!")
                pygame.quit()
            
            # Movimentos do Monstro
            dire = [0,0,0,0]
            for i in range(len(monster)):
                
                # Perseguição com limitações de tela
                # Vertical: quando a posição [0] é a mesma
                if monster[i].rect[0] == player.rect[0]: 
                    # UP
                    if monster[i].rect[1] >  player.rect[1]: 
                        monster[i].rect[1] -= 50  
                    # DOWN
                    if monster[i].rect[1] <  player.rect[1]: 
                        monster[i].rect[1] += 50 
                    # LOSE
                    if(monster[i].rect[0] == player.rect[0] and
                       monster[i].rect[1] == player.rect[1]): 
                        print("Você foi pego!!!")
                        pygame.quit()
                
                # Horizontal: quando a altura [1] é a mesma
                elif monster[i].rect[1] == player.rect[1]:
                    # WALL
                    if(monster[i].rect[1] < 250 or monster[i].rect[1] > 450):
                        dire[i] = randint(0,5)
                        # UP
                        if(dire[i] == 1 and 
                           monster[i].rect[1] >  50 and 
                           monster[i].rect[0] != 500):  
                            monster[i].rect[1] -= 50 
                        # DOWN
                        if(dire[i] == 2 and 
                           monster[i].rect[1] < 600 and 
                           monster[i].rect[0] != 500):  
                            monster[i].rect[1] += 50 
                        # LEFT
                        if(dire[i] == 3 and 
                           monster[i].rect[0] >  50 and 
                           monster[i].rect[0] != 550):  
                            monster[i].rect[0] -= 50 
                        # RIGHT
                        if(dire[i] == 4 and 
                           monster[i].rect[0] < 1100 and
                           monster[i].rect[0] != 450): 
                            monster[i].rect[0] += 50 
                    # WITHOUT WALL
                    if(monster[i].rect[1] > 250 and monster[i].rect[1] < 450):
                        # LEFT
                        if(monster[i].rect[0] > player.rect[0]):
                            monster[i].rect[0] -= 50
                        # RIGHT
                        if(monster[i].rect[0] < player.rect[0]):
                            monster[i].rect[0] += 50
                    # LOSE             
                    if(monster[i].rect[0] == player.rect[0] and
                       monster[i].rect[1] == player.rect[1]): 
                        print("Você foi pego!!!")
                        pygame.quit()
                
                else:
                # movimentos aleatórios 
                # com limitações de tela
                    dire[i] = randint(0,5)
                    # UP
                    if(dire[i] == 1 and 
                       monster[i].rect[1] >  50 and 
                       monster[i].rect[0] != 500):  
                        monster[i].rect[1] -= 50 
                    # DOWN
                    if(dire[i] == 2 and 
                       monster[i].rect[1] < 600 and 
                       monster[i].rect[0] != 500):  
                        monster[i].rect[1] += 50 
                    # LEFT
                    if(dire[i] == 3 and 
                       monster[i].rect[0] >  50 and 
                       monster[i].rect[0] != 550):  
                        monster[i].rect[0] -= 50 
                    # RIGHT
                    if(dire[i] == 4 and 
                       monster[i].rect[0] < 1100 and
                       monster[i].rect[0] != 450): 
                        monster[i].rect[0] += 50 
            
            # WIN !!
            if(player.rect[0] == star[0].rect[0] and
               player.rect[1] == star[0].rect[1]): 
                print("""\n
                         \        /  0   |\    |    ||  
                          \  /\  /   |   |  \  |    ||
                           \/  \/    |   |    \|    00""")
                pygame.quit()
            
            # imagens e atualização
            fundo.fill((255,255,255))
            
            player_group.update()
            monster_group.update()
            box_group.update()
            star_group.update()

            star_group.draw(fundo)
            box_group.draw(fundo)
            monster_group.draw(fundo)
            player_group.draw(fundo)

            fundo.blit(texto5, pos_texto5)
    
            pygame.display.update()

jogo()
pygame.quit()
```
