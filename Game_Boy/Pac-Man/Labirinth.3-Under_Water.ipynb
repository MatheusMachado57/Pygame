```python
import pygame, random
from pygame.locals import *
from random import randint

SCREEN_WIDTH = 1200
SCREEN_HEIGHT = 700

fundo = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

class Player(pygame.sprite.Sprite):

    def __init__(self):
        pygame.sprite.Sprite.__init__(self)
        self.images = [pygame.image.load('vermelhoD.png').convert_alpha(),
                       pygame.image.load('vermelhoE.png').convert_alpha(),
                       pygame.image.load('vermelhoC.png').convert_alpha(),
                       pygame.image.load('vermelhoB.png').convert_alpha()]
        self.current_image = 0
        self.image = pygame.image.load('vermelhoB.png').convert_alpha()
        self.rect = self.image.get_rect()
        self.rect[0] = 550
        self.rect[1] = 350
            
    def go_up(self):   self.image = pygame.image.load('vermelhoC.png').convert_alpha()     
    def go_down(self): self.image = pygame.image.load('vermelhoB.png').convert_alpha()
    def go_on(self):   self.image = pygame.image.load('vermelhoD.png').convert_alpha()
    def go_back(self): self.image = pygame.image.load('vermelhoE.png').convert_alpha()            
                
# Paredes e Limitações:           
w1 = []

# Horizontal:
for i in range( 4,24): w1.append([i*50 + 450,   50 + 400])
for i in range(23):    w1.append([i*50 + 450,  650 + 400])
for i in range( 1,6):  w1.append([i*50 + 400, -150 + 400])
for i in range( 4,23): w1.append([i*50 + 450,  200 + 400])    
for i in range( 4,23): w1.append([i*50 + 450,  350 + 400])
for i in range( 4,23): w1.append([i*50 + 450,  500 + 400])    
for i in range( 4,23): w1.append([i*50 + 450,  650 + 400])

# Vertical: 
for i in range(11,14): w1.append([13*50, i*50 - 250])
for i in range(11,14): w1.append([ 9*50, i*50 - 250])
    
# Box:
box = []
Veloc = 50                
    
for i in range(len(w1)):
    class Box(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('preto.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('preto.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w1[i][0]
            self.rect[1] = w1[i][1]

        def update(self): 
            self.image = self.images[self.current_image]
            
        def go_up(self):   self.rect[1] += Veloc
        def go_down(self): self.rect[1] -= Veloc
        def go_on(self):   self.rect[0] -= Veloc
        def go_back(self): self.rect[0] += Veloc
            
    box.append(Box())
# Preparo das aparências do jogo:  
pygame.init()
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

# Jogador
player_group = pygame.sprite.Group()
player = Player()
player_group.add(player)

# Limitações:
box_group = pygame.sprite.Group()
box_group.add(box)

clock = pygame.time.Clock()    

def jogo():     
    # tempo, moedas e corações:
    font = pygame.font.SysFont('arial black', 20)

    # informações essenciais:
    texto5 = font.render(" Para sair do jogo aperte q ", True, (255,255,255), (0,0,0))
    pos_texto5 = texto5.get_rect()
    pos_texto5.center = (200,20)
    
    # parte original    
    play = True
    while play:
        
        # regulador de velocidade de jogo: 
        clock.tick(10)

        # caso se feche a janela:
        for event in pygame.event.get(): 
            if event.type == QUIT: pygame.quit()
                            
        # se ainda houver vidas ou continue:
        else:
            # comandos do jogador:    
            comandos = pygame.key.get_pressed()
            
            if comandos[pygame.K_UP]:    
                M = True
                player.go_up()
                for i in range(len(box)): 
                    if (player.rect[1]  - box[i].rect[1] == Veloc and
                        player.rect[0] == box[i].rect[0]): M = False
                if M == True: 
                    for i in range(len(box)): box[i].go_up()
            
            if comandos[pygame.K_DOWN]:
                M = True
                player.go_down()
                for i in range(len(box)): 
                    if (box[i].rect[1] - player.rect[1] == Veloc and
                        player.rect[0] == box[i].rect[0]): M = False
                if M == True: 
                    for i in range(len(box)): box[i].go_down()
            
            if comandos[pygame.K_LEFT]:  
                M = True
                player.go_back()
                for i in range(len(box)): 
                    if (player.rect[0] -  box[i].rect[0] == Veloc and
                        player.rect[1] == box[i].rect[1]): M = False
                if M == True: 
                    for i in range(len(box)): box[i].go_back()
            
            if comandos[pygame.K_RIGHT]: 
                M = True
                player.go_on()
                for i in range(len(box)): 
                    if (box[i].rect[0] - player.rect[0] == Veloc and
                        player.rect[1] == box[i].rect[1]): M = False
                if M == True: 
                    for i in range(len(box)): box[i].go_on()
            
            if comandos[pygame.K_q]:
                print("U Quit!!!")
                pygame.quit()
            
            # imagens e atualização
            # azul
            fundo.fill((0,50,255)) # 255
            
            player_group.update()
            box_group.update()

            box_group.draw(fundo)
            player_group.draw(fundo)

            fundo.blit(texto5, pos_texto5)
    
            pygame.display.update()

jogo()
pygame.quit()
```
