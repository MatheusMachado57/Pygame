```python
import pygame, sys, time
from pygame.locals import *

# set up pygame
pygame.init()

# set up the window
largura = 400
altura  = 400
windowSurface = pygame.display.set_mode((largura, altura), 0, 32)

branco    = (255,255,255)
cinza     = (127,127,127)
verde     = (128,255,0)
azul      = (100,159,237)
vermesc   = (127,0,0)
rosa      = (255,200,200)
preto     = (0,0,0)
amarelo   = (255,255,0)
amarclar  = (255,255,151)
marrom    = (128,0,0)

pygame.display.set_caption("Montanha")

def Montanha():
     
    sair = True 
    while sair: 
        
        for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    sair = False  
        
        windowSurface.fill(branco)
                    
        # Montanha de subida diagonal:
        ## 1 Horizontal:
        l = []
        a = []
        px = 50
        py = 50
        t  = 2
        
        q1 = 180
        for i in range(q1):
            b = {"rect":pygame.Rect(px, py, t, t), "color":preto}
            a.append(b)
            l.append([px,py])
            py += t     
        
        circ = 0
        for i in range(25):
            
            circ += 2 
            px = px + t
            py = 50 - circ  
            
            for i in range(1):
                b = {"rect":pygame.Rect(px, py, t, t), "color":preto}
                a.append(b)
                l.append([px,py])
                py += t 
            
            q1 = q1 + t
            
            for i in range(q1):
                b = {"rect":pygame.Rect(px, py, t, t), "color":verde}
                a.append(b)
                l.append([px,py])
                py += t 
            
        for i in range(len(a)):
            pygame.draw.rect(windowSurface, a[i]["color"], a[i]["rect"])  
            
        pygame.display.update()

Montanha()
pygame.quit()
```
