```python
import pygame, time
from pygame.locals import *
from random import randint

pygame.init()

largura = 1200 
altura  = 800

janela = pygame.display.set_mode((1300, 700))
pygame.display.set_caption('Barrel')

def jogo():
    
    sair = True
    fimdejogo = False
    
    pygame.time.delay(60)     # dá um delay necessário para vermos o desenho;  

    # posição no eixo x e no eixo y;
    pos_y0 = altura - 220                     
    pos_x0 = 100                                          

    pos_x1 = 1000
    pos_y1 = 100       
    
    velocB = 20 
    veloc1 = 10
    veloc2 = 40
    
    descendo = True
    l = 1
    
    fundo = pygame.image.load('Boss.PNG')

    b1 = pygame.image.load('voando.PNG')
    b2 = pygame.image.load('voando - Copia.PNG')
    bonec = b1

    f1 = pygame.image.load('crashed.png')
    
    while sair:
                              
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                sair = False
        
        # Gravidade:
        if pos_y0 < altura - 220:
            pos_y0 += velocB
            
        # movimentos e limitações;    
        comandos = pygame.key.get_pressed()
        
        if comandos[pygame.K_UP] and pos_y0 > 400:      # jump
                pos_y0 -= 3*velocB
             
        if comandos[pygame.K_LEFT] and pos_x0 > 0:
                bonec = b2
                pos_x0 -= velocB
                
        if comandos[pygame.K_RIGHT] and pos_x0 < 1180:
                bonec = b1
                pos_x0 += velocB   
                
        # Movimentação:
       
        # Lugar de descida 1: ######################################3
        if l == 1:         
            # descendo:
            if pos_y1 <= 400 and descendo == True: 
                pos_y1 += veloc2
            else:
                descendo = False
            
            # subindo:
            if pos_y1 >= 100 and descendo == False: 
                pos_y1 -= veloc1
                if pos_y1 <= 100:     # se if pos_y1 >= 100: o barril desce, muda de lugar por baixo e sobe;
                    l += 1
                    descendo = True
           #else:
               #descendo = True       # se fizermos isso, o barril sobe e desce sem parar;         
        
        # Lugar de descida 2: ######################################3
        if (l == 2) and (pos_x1 > 500):
            pos_x1 -= veloc1
        
        if (l == 2) and (pos_x1 <= 500):
            # descendo:
            if pos_y1 <= 400 and descendo == True: 
                pos_y1 += veloc2
            else:
                descendo = False
            
            # subindo:
            if pos_y1 >= 100 and descendo == False: 
                pos_y1 -= veloc1
                if pos_y1 <= 100:
                    l += 1
                    descendo = True
#            else:
#                l = 2
            #     descendo = True
            
        # Lugar de descida 3: ######################################3
        if (l == 3) and pos_x1 > 100:
            pos_x1 -= veloc1
        
        if(l == 3) and (pos_x1<=100):
            # descendo:
            if pos_y1 <= 400 and descendo == True: 
                pos_y1 += veloc2
            else:
                descendo = False
            
            # subindo:
            if pos_y1 >= 100 and descendo == False: 
                pos_y1 -= veloc1
                if pos_y1 <=100:
                    l += 1
#            else:
#                l = 3                  
            #     descendo = True  
        
        # Voltando ao lugar de descida 1: 
        if (l == 4):
            pos_x1 += veloc1
            if pos_x1 >= largura - 200:
                l -= 3
                descendo = True
        
        
        # invocação das figuras;        
        janela.blit(fundo, (0, 0))
        janela.blit(bonec, (pos_x0, pos_y0))        
        janela.blit(f1, (pos_x1, pos_y1))
                   
        pygame.display.update()                                    # atualiza a tela

jogo()        
pygame.quit()                        
```
