```python
import sys

try:
    import Tkinter as tk
except ImportError:
    import tkinter as tk

try:
    import ttk
    py3 = False
except ImportError:
    import tkinter.ttk as ttk
    py3 = True

############################################################################

import pygame, random
from pygame.locals import *

SCREEN_WIDTH = 1000
SCREEN_HEIGHT = 600

GROUND_WIDTH = SCREEN_WIDTH
GROUND_HEIGHT = SCREEN_HEIGHT - 30

fundo = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

moedas_p = 0
vidas_p  = 0

class Ground(pygame.sprite.Sprite):
    def __init__(self):
        pygame.sprite.Sprite.__init__(self)

        self.image = pygame.image.load('layer-5.png').convert_alpha()
        self.image = pygame.transform.scale(self.image, (SCREEN_WIDTH, SCREEN_HEIGHT))
        self.mask = pygame.mask.from_surface(self.image)
        self.rect = self.image.get_rect()

class Player(pygame.sprite.Sprite):

    def __init__(self, isjump):
        pygame.sprite.Sprite.__init__(self)
        self.images = [pygame.image.load('azul.png').convert_alpha()]
        self.current_image = 0
        self.image = pygame.image.load('azul.png').convert_alpha()
        self.rect = self.image.get_rect()
        self.rect[0] = 10
        self.rect[1] = GROUND_HEIGHT - 100
        self.fall = False
        self.vidas_p = 1
        
    def update(self):
        self.image = self.images[self.current_image]
        
        if pygame.sprite.groupcollide(player_group, fire_group, True, False, pygame.sprite.collide_mask):
            self.vidas_p -= 1
            
        if player.rect[1] - croc[0].rect[1] < 40:
            if pygame.sprite.groupcollide(player_group, croc_group, True, False, pygame.sprite.collide_mask):
                    self.vidas_p -= 1

        # Update height
        if (pygame.sprite.groupcollide(player_group, ground_group, False, False, pygame.sprite.collide_mask) or
            pygame.sprite.groupcollide(player_group, box_group, False, False, pygame.sprite.collide_mask) or
            isjump):  
            return
        
        # gravidade ou não
        if self.rect[1] > GROUND_HEIGHT - 100: 
            self.rect[1] -= 5
            self.fall = False
        else: 
            self.rect[1] += 5
            self.fall = True
            
    def go_on(self): 
        if self.rect[0] < 900: 
            self.rect[0] += 5
        
        for i in range(len(box)):
            if abs(self.rect[0] - box[i].rect[0]) < 50 and abs(self.rect[1] - box[i].rect[1]) < 40:
                 self.rect[0] -= 5
    
    def go_back(self): 
        if self.rect[0] > 0: 
            self.rect[0] -= 5
            
        for i in range(len(box)):    
            if abs(self.rect[0] - box[i].rect[0]) < 50 and abs(self.rect[1] - box[i].rect[1]) < 40:
                self.rect[0] += 5
                
# Inimigo: ##################################################################

w =[[800, GROUND_HEIGHT - 100]] 
croc = []
                
for i in range(len(w)):
    class Croc(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('verde.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('verde.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w[i][0]
            self.rect[1] = w[i][1]

        def update(self):
            self.image = self.images[self.current_image] 
            pygame.sprite.groupcollide(player_group, croc_group, False, True, pygame.sprite.collide_mask)
            
            
    croc.append(Croc())

# Objeto inanimado: #########################################################3               
                
w =[[300, GROUND_HEIGHT - 100],
    [350, GROUND_HEIGHT - 150],
    [350, GROUND_HEIGHT - 100],
    [650, GROUND_HEIGHT - 100],
    [650, GROUND_HEIGHT - 150],
    [700, GROUND_HEIGHT - 100]] 
box = []
                
for i in range(len(w)):
    class Box(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('marrom.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('marrom.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w[i][0]
            self.rect[1] = w[i][1]

        def update(self):
            self.image = self.images[self.current_image]
            
    box.append(Box())
    
# Moedas: #########################################################################    
        
w = [[100, 200],
     [120, 200],
     [140, 200]]      
coin = []        
        
for i in range(len(w)):
    class Coin(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('amarela.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('amarela.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w[i][0]
            self.rect[1] = w[i][1]

        def update(self):
            self.image = self.images[self.current_image]
            if pygame.sprite.groupcollide(player_group, coin_group, False, True, pygame.sprite.collide_mask):
                moedas_p += 1
            
            
    coin.append(Coin())
    
# Coração: ###############################################################################
        
w = [[100, 220],
     [120, 220],
     [140, 220]]
coração = []

for i in range(len(w)):
    class Coração(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('vermelho.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('vermelho.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w[i][0]
            self.rect[1] = w[i][1]

        def update(self):
            self.image = self.images[self.current_image]
            if pygame.sprite.groupcollide(player_group, coração_group, False, True, pygame.sprite.collide_mask):
                vidas_p += 1
            
    coração.append(Coração())
    
# Fogo: ##############################################################################################
w = []
j = 400
for i in range(25):
    w.append([j, 510])
    j += 10

fire = []

for i in range(len(w)):
    class Fire(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('laranja.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('laranja.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w[i][0]
            self.rect[1] = w[i][1]

        def update(self):
            self.image = self.images[self.current_image]
            pygame.sprite.groupcollide(player_group, coração_group, False, True, pygame.sprite.collide_mask)
            
    fire.append(Fire())

# Preparo das aparências do jogo: ####################################################################   
pygame.init()
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

# Terreno
ground_group = pygame.sprite.Group()
ground = Ground()
ground_group.add(ground)

# Jogador
isjump = False
player_group = pygame.sprite.Group()
player = Player(isjump)
player_group.add(player)

# Inimigo:
croc_group = pygame.sprite.Group()
croc_group.add(croc)

# Objeto Inanimado
box_group = pygame.sprite.Group()
box_group.add(box)

# Moedas
coin_group = pygame.sprite.Group()
coin_group.add(coin)

# Coração
coração_group = pygame.sprite.Group()
coração_group.add(coração)

# Fire
fire_group = pygame.sprite.Group()
fire_group.add(fire)

clock = pygame.time.Clock()    

# Começo: ##################################

class Toplevel1:
    def __init__(self):
        '''This class configures and populates the toplevel window.
           top is the toplevel containing window.'''
        _bgcolor = '#d9d9d9'  # X11 color: 'gray85'
        _fgcolor = '#000000'  # X11 color: 'black'
        _compcolor = '#d9d9d9' # X11 color: 'gray85'
        _ana1color = '#d9d9d9' # X11 color: 'gray85'
        _ana2color = '#ececec' # Closest X11 color: 'gray92'
        font10 = "-family {Segoe UI} -size 15"
        font9 = "-family {Segoe UI} -size 20"
        
        self.root = tk.Tk()        
        self.root.geometry("600x183+402+159")
        self.root.minsize(120, 1)
        self.root.maxsize(1370, 749)
        self.root.resizable(1, 1)
        self.root.title("New Toplevel")
        self.root.configure(background="#808080")
        self.root.configure(highlightcolor="#646464")

        self.Continue = tk.Frame(self.root)
        self.Continue.place(relx=0.05, rely=0.089, relheight=0.656, relwidth=0.625)

        self.Continue.configure(relief='groove')
        self.Continue.configure(borderwidth="2")
        self.Continue.configure(relief="groove")
        self.Continue.configure(background="#0000a0")
        self.Continue.configure(highlightbackground="#000080")
        self.Continue.configure(highlightcolor="#0000a0")

        self.Message1 = tk.Message(self.Continue)
        self.Message1.place(relx=0.317, rely=0.158, relheight=0.202,relwidth=0.45)
        self.Message1.configure(background="#d9d9d9")
        self.Message1.configure(borderwidth="5")
        self.Message1.configure(font=font9)
        self.Message1.configure(foreground="#000000")
        self.Message1.configure(highlightbackground="#d9d9d9")
        self.Message1.configure(highlightcolor="black")
        self.Message1.configure(text='''Again ???''')
        self.Message1.configure(width=190)

        self.Button1 = tk.Button(self.Continue, self.begin)
        self.Button1.place(relx=0.233, rely=0.437, height=64, width=107)
        self.Button1.configure(activebackground="#ececec")
        self.Button1.configure(activeforeground="#000000")
        self.Button1.configure(background="#0000ff")
        self.Button1.configure(disabledforeground="#a3a3a3")
        self.Button1.configure(font=font10)
        self.Button1.configure(foreground="#000000")
        self.Button1.configure(highlightbackground="#d9d9d9")
        self.Button1.configure(highlightcolor="black")
        self.Button1.configure(pady="0")
        self.Button1.configure(text='''Yes''')
        
        self.Button2 = tk.Button(self.Continue, self.over)                          
        self.Button2.place(relx=0.533, rely=0.437, height=64, width=107)
        self.Button2.configure(activebackground="#ececec")
        self.Button2.configure(activeforeground="#000000")
        self.Button2.configure(background="#ff0000")
        self.Button2.configure(disabledforeground="#a3a3a3")
        self.Button2.configure(font=font10)
        self.Button2.configure(foreground="#000000")
        self.Button2.configure(highlightbackground="#d9d9d9")
        self.Button2.configure(highlightcolor="black")
        self.Button2.configure(pady="0")
        self.Button2.configure(text='''No''')
        
        self.root.mainloop()

    def begin():
    
        # Marcador de tempo:
        timer = 0
        tempo_segundos = 0
        font = pygame.font.SysFont('arial black', 15)
            
        #texto1 = font.render("Tempo:", True, (0,0,0), (255,255,255))
        #pos_texto1 = texto1.get_rect()
        #pos_texto1.center = (40,40) 
            
        texto2 = font.render("Coração:", True, (0,0,0), (255,255,255))
        pos_texto2 = texto2.get_rect()
        pos_texto2.center = (40,65) 
            
        texto3 = font.render("Moedas:", True, (0,0,0), (255,255,255))
        pos_texto3 = texto3.get_rect()
        pos_texto3.center = (40,90)
            
        # variáveis da desgraça do pulo:
        isjump = False
        jumpcount = 20
        v = []
        while jumpcount >= -20:    
            if jumpcount < 0: neg = -1
            else: neg = 1
            v.append((jumpcount**2)*0.5*neg/10)
            jumpcount -= 1

        jumpcount = 20
        i = 0
        j = 0
        s = []
    
        while True:
            clock.tick(70)
    
            for event in pygame.event.get():
                if event.type == QUIT:
                    pygame.quit()
    
            # comandos do jogador:    
            comandos = pygame.key.get_pressed()
            if  comandos[pygame.K_LEFT]:  player.go_back()
            if comandos[pygame.K_RIGHT]: player.go_on()  
    
            # pulo
            if not(isjump):
                if  ((comandos[pygame.K_SPACE] and 
                    (ground.rect[1] - player.rect[1] < -469 or 
                     pygame.sprite.groupcollide(player_group, box_group, False, False, pygame.sprite.collide_mask))) or
                     pygame.sprite.groupcollide(player_group, croc_group, False, True, pygame.sprite.collide_mask)):            
                    isjump = True
                    s.append(v)
            else:
                if jumpcount >= -20: 
                    if jumpcount < 0: neg = -1
                    else: neg = 1    
                    player.rect[1] -= s[j][i]
                    if pygame.sprite.groupcollide(player_group, box_group, False, False, pygame.sprite.collide_mask):
                         player.rect[1] += s[j][i]
                    if pygame.sprite.groupcollide(player_group, croc_group, False, False, pygame.sprite.collide_mask):
                        i = 0
                        s[0] = v
                        jumpcount = 20
                    jumpcount -= 1
                    i += 1
                else: 
                    i = 0
                    s.remove(v)
                    jumpcount = 20
                    isjump = False
    
            if moedas_p == 100:
                    mordas_p = 0
                    player.vidas_p += 1
    
            if player.vidas_p == 0:
                    Toplevel1()
    
                # marcador de tempo:
#                if(timer < 20): timer += 1
#                else:
#                    tempo_segundos += 1
#                    texto1 = font.render("  Tempo: " + str(tempo_segundos)+ " ", True, (0,0,0), (225,225,255))
    
            # marcador de vida:
            texto2 = font.render("  Corações: " + str(int(player.vidas_p))+ " ", True, (0,0,0), (225,225,255))
    
            # marcador de moedas:
            texto3 = font.render("  Moedas: " + str(moedas_p)+ " ", True, (0,0,0), (225,225,255))
    
            # imagens e atualização
            fundo.fill((127,127,127))
#           fundo.blit(texto1, pos_texto1)
            fundo.blit(texto2, pos_texto2)
            fundo.blit(texto3, pos_texto3)
  
            player_group.update()
            box_group.update()
            coin_group.update()
            coração_group.update()
            croc_group.update()
            fire_group.update()
    
            fire_group.draw(screen)
            croc_group.draw(screen)
            coração_group.draw(screen)
            coin_group.draw(screen)
            box_group.draw(screen)
            player_group.draw(screen)
    
            ground_group.draw(screen)
            pygame.display.update()

    def over():
         pygame.quit()
         
Toplevel1.begin()
```
