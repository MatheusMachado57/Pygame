```python
import pygame, random
from pygame.locals import *

SCREEN_WIDTH = 1000
SCREEN_HEIGHT = 600

GROUND_WIDTH = SCREEN_WIDTH
GROUND_HEIGHT = SCREEN_HEIGHT - 30

fundo = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

class Ground(pygame.sprite.Sprite):
    def __init__(self):
        pygame.sprite.Sprite.__init__(self)

        self.image = pygame.image.load('layer-5.png').convert_alpha()
        self.image = pygame.transform.scale(self.image, (SCREEN_WIDTH, 600))
        self.mask = pygame.mask.from_surface(self.image)
        self.rect = self.image.get_rect()

class Player(pygame.sprite.Sprite):

    def __init__(self):
        pygame.sprite.Sprite.__init__(self)
        self.images = [pygame.image.load('azul.png').convert_alpha()]
        self.current_image = 0
        self.image = pygame.image.load('azul.png').convert_alpha()
        self.rect = self.image.get_rect()
        self.rect[0] = 10
        self.rect[1] = GROUND_HEIGHT - 100
        self.fall = False
        self.vidas_p = 5
        
    def update(self):
        
        # Calibrando a gravidade:
        player.rect[1] += 1
        
        for i in range(len(box)):
             if (player.rect[1] >= GROUND_HEIGHT - 100 or
                 isjump == True or
                (abs(player.rect[1] - box[i].rect[1]) < 50 and 
                 abs(player.rect[0] - box[i].rect[0]) < 50)):
                    player.rect[1] -= 1
        
    def go_on(self): 
        if self.rect[0] < 900: 
            self.rect[0] += 5
        
        for i in range(len(box)):
            if abs(self.rect[0] - box[i].rect[0]) < 50 and abs(self.rect[1] - box[i].rect[1]) < 46:
                 self.rect[0] -= 5
    
    def go_back(self): 
        if self.rect[0] > 0: 
            self.rect[0] -= 5
            
        for i in range(len(box)):    
            if abs(self.rect[0] - box[i].rect[0]) < 50 and abs(self.rect[1] - box[i].rect[1]) < 46:
                self.rect[0] += 5
                
# Inimigo: ##################################################################

w =[[300, GROUND_HEIGHT - 100],
    [500, GROUND_HEIGHT - 100],
    [700, GROUND_HEIGHT - 100]] 

croc = []
                
for i in range(len(w)):
    class Croc(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('verde.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('verde.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w[i][0]
            self.rect[1] = w[i][1]

        def update(self):
            self.image = self.images[self.current_image] 
            
    croc.append(Croc())

# Objeto inanimado: #########################################################3               
                
w =[[200, GROUND_HEIGHT - 100],
    [400, GROUND_HEIGHT - 100],
    [400, GROUND_HEIGHT - 150],
    [600, GROUND_HEIGHT - 100],
    [800, GROUND_HEIGHT - 100]] 

box = []
                
for i in range(len(w)):
    class Box(pygame.sprite.Sprite):
        def __init__(self):
            pygame.sprite.Sprite.__init__(self)
            self.images = [pygame.image.load('marrom.png').convert_alpha()]
            self.current_image = 0
            self.image = pygame.image.load('marrom.png').convert_alpha()
            self.mask = pygame.mask.from_surface(self.image)
            self.rect = self.image.get_rect()
            self.rect[0] = w[i][0]
            self.rect[1] = w[i][1]

        def update(self):
            self.image = self.images[self.current_image]
            
    box.append(Box())
    
# Preparo das aparências do jogo: ####################################################################   
pygame.init()
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))

# Terreno
ground_group = pygame.sprite.Group()
ground = Ground()
ground_group.add(ground)

# Jogador
player_group = pygame.sprite.Group()
player = Player()
player_group.add(player)

# Inimigo:
croc_group = pygame.sprite.Group()
croc_group.add(croc)

# Objeto Inanimado
box_group = pygame.sprite.Group()
box_group.add(box)

clock = pygame.time.Clock()    
isjump = False

def jogo(): 
    
    # tempo, moedas e corações:
    font = pygame.font.SysFont('arial black', 15)

    texto2 = font.render("Coração:", True, (0,0,0), (255,255,255))
    pos_texto2 = texto2.get_rect()
    pos_texto2.center = (40,65) 

    ################################
    
    # variáveis da desgraça do pulo:
    jumpcount = 20
    v = []    
    while jumpcount >= -20:    
        if jumpcount < 0: neg = -1
        else: neg = 1
        v.append((jumpcount**2)*0.5*neg/7)
        jumpcount -= 1
    v.append(-19.5)
    jumpcount = 20
    i = 0
    j = 0
    s = []
    isjump = False
    ###############################
    
    # parte original    
    fimdejogo = False
    play = True
    while play:
        
        # regulador de velocidade de jogo: 
        clock.tick(50)
        #################################
        
        # caso se feche a janela:
        for event in pygame.event.get():
            if event.type == QUIT:
                pygame.quit()
        #################################
       
        # game over!!
        if player.vidas_p == 0: fimdejogo = True
        if fimdejogo == True: 
            return
        
        #se ainda houver vidas ou continue:
        else:
    
            # comandos do jogador:    
            comandos = pygame.key.get_pressed()
            if comandos[pygame.K_LEFT]:  player.go_back()
            if comandos[pygame.K_RIGHT]: player.go_on()     
             
            # pulo
            if not(isjump):
                for q in range(len(box)):
                    if  (comandos[pygame.K_SPACE] and  
                        (abs(player.rect[1] - box[q].rect[1]) < 50 or
                         player.rect[1] >= SCREEN_HEIGHT - 130)):
                          isjump = True
            else:
                    s.append(v)
                    if jumpcount >= -20: 
                        if jumpcount < 0: neg = -1
                        else: neg = 1
                        #    
                        for r in range(len(box)):    
                            if (abs(player.rect[1] - box[r].rect[1]) < 51 and abs(player.rect[0] - box[r].rect[0]) < 50):
                                 player.rect[1] += s[j][i]
                        #        
                        if pygame.sprite.groupcollide(player_group, croc_group, False, False, pygame.sprite.collide_mask):
                                i = 0
                                s[0] = v
                                jumpcount = 20
                        #        
                        player.rect[1] -= s[j][i]
                        jumpcount -= 1
                        i += 1
                    else:
                        i = 0
                        s.remove(v)
                        jumpcount = 20
                        isjump = False
            # Fim do pulo
            
            # marcador de vida:
            texto2 = font.render("  Corações: " + str(int(player.vidas_p))+ " ", True, (0,0,0), (225,225,255))
    
            # imagens e atualização
            fundo.fill((127,127,127))
            fundo.blit(texto2, pos_texto2)
  
            player_group.update()
            box_group.update()
            croc_group.update()
    
            croc_group.draw(fundo)
            box_group.draw(fundo)
            player_group.draw(fundo)
    
            ground_group.draw(fundo)
            pygame.display.update()

jogo()
pygame.quit()
```
