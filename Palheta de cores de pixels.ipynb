```python
import pygame, sys
from pygame.locals import *

# set up pygame
pygame.init()

# set up the window
largura = 1200
altura  = 200
windowSurface = pygame.display.set_mode((largura, altura), 0, 32)

branco    = (255,255,255)
cinza     = (127,127,127)
preto     = (0  ,0  ,  0)
prateado  = (192,192,192)

azul      = (0  ,  0,255)
azulclaro = (100,149,237)
azulescur = (0  ,0  ,128)
azulpetró = (0  ,128,128)
azulpisc  = (0  ,255,255)

azulpisc2 = (0,200,255)

fúcsia    = (255,  0,255)    # rosa escuro

marrom    = (128,  0,  0)
marrom2 =   (200,100,0)

roxo      = (128,  0,128)
amarelo   = (255,255,  0)

verde     = (0  ,128,  0)
verdeclr  = (0  ,255,  0)
verdeoliv = (128,128,  0)

rosa = (255, 100, 255)
vinho = (200,0,128)
vermelho  = (255,0  ,  0)

laranja = (255, 100, 0)
laranja2 = (255, 175, 0)

rosac = (255,200,255)
rosac2 = (255,200,200)
amaresc = (200,180,0)

amarpok = (240,220,0)

cores = [branco, cinza, preto, prateado, vermelho,
         azul, azulclaro, azulescur, azulpetró,
         azulpisc, fúcsia, marrom, roxo, amarelo,
         verde, verdeclr, verdeoliv, rosa, vinho,
         amaresc, rosac, azulpisc2, marrom2, rosac2, 
         laranja, laranja2, amarpok]

pygame.display.set_caption("Cores")

def Cores():
     
    sair = True 
    while sair: 
        
        for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    sair = False  
        
        windowSurface.fill(branco)
                    
        # Cores:
        a = []
        px = 10
        py = 100
        t  = 20
        
        ## 1 Horizontal:
        for i in range(len(cores)):
            b = {"rect":pygame.Rect(px, py, t, t), "color":cores[i]}
            a.append(b)
            px += t*2    
        
        for i in range(len(a)):
            pygame.draw.rect(windowSurface, a[i]["color"], a[i]["rect"])  
            
        pygame.display.update()

Cores()
pygame.quit()
```
